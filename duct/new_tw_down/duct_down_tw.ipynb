# Нижняя ветвь "новой" бегущей волны в канале

Была найдена "новая" бегущая волна, выделенная в плоском канале непосредственно из турбулентного течения. Лучшего названия я для него пока не придумал. Это решение не очень удобно для исследования так как амплитуда пульсаций в нем велика. 

В одной из статей было отмечено, что в каждом решении, которое они рассматривали, верхняя ветвь оказывается внутри области притяжения турбулентности в то время, как нижняя ветвь принадлежит ее границе, то есть сепаратрисе. Соответственно, только верхняя ветвь решения может находится внутри турбулентного аттрактора. 

Оказалось, что найденная бегущая волна действительно принадлежит верхней ветви решения. Продлевая его в сторону уменьшения числа Рейнольдса до точки бифуркации удалось получить решение на нижней ветви (Рис.1.).

Решение на нижней ветви имеет значительно более низкую амплитуду пульсаций, чем на верхней. Это делает его идеальным кандидатом для исследования: решения является пристенным, система координат афинная, пульсации имеют небольшую величину. 

При продлении решения в пространстве параметров все параметры, кроме Re и cf, оставались постоянными

## Механизм возникновения ox по бокам от полосы замедления

Для того, чтобы выяснить причину возникновения пульсационной составляющей продольной завихренности o_x по бокам от полосы замедления, обратимся к уравнениям. 

Линеаризованные уравнения на пульсационную составляющею продольной завихренности o_x

$$
\frac{\partial \omega_x}{\partial t} = c_f \frac{\partial \omega_x}{\partial x} - rot_x[\Omega \times u] - rot_x[\omega \times U]  + \nu \Delta \omega_x
$$

В этом разделе большие буквы - среднее течение, маленькие - пульсационное, штрихи опущены.


$$
 - rot_x[\Omega \times u] = - u \frac{\partial \Omega_x}{\partial x} - v \frac{\partial \Omega_x}{\partial y} - w \frac{\partial \Omega_x}{\partial z} + \Omega_x \frac{\partial u}{\partial x} + \Omega_y \frac{\partial u}{\partial y} + \Omega_z \frac{\partial u}{\partial z}
$$

$$
c_f \frac{\partial \omega_x}{\partial x} - rot_x[\omega \times U] = - (U - c_f) \frac{\partial \omega_x}{\partial x} - V \frac{\partial \omega_x}{\partial y} - W \frac{\partial \omega_x}{\partial z} + \omega_x \frac{\partial U}{\partial x} + \omega_y \frac{\partial U}{\partial y} + \omega_z \frac{\partial U}{\partial z}
$$



Как и раньше, разделим все слагаемые на три группы:

Растяжение и смещение вдоль трубы 
$$
с_1 = - u \frac{\partial \Omega_x}{\partial x} - (U - c_f) \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial U}{\partial x} + \Omega_x \frac{\partial u}{\partial x}
$$

Конвекция в поперечной плоскости
$$
c_2 = - v \frac{\partial \Omega_x}{\partial y} - w \frac{\partial \Omega_x}{\partial z} - V \frac{\partial \omega_x}{\partial y} - W \frac{\partial \omega_x}{\partial z}
$$

Поворот существующих вихревых линий
$$
c_3 = \Omega_y \frac{\partial u}{\partial y} + \Omega_z \frac{\partial u}{\partial z} + \omega_y \frac{\partial U}{\partial y} + \omega_z \frac{\partial U}{\partial z}
$$

На следующем рисунке приведен вклад в энергию продольной завихренности o_x от каждого из слагаемых c1,c2,c3.

<img src="ox_gen.png">
Рис.16. Вклад в энергию продольной завихренности o_x пульсационной составляющей движения, осредненный вдоль потока. Первая картинка -  суммарный вклад, вторая - c1\*o_x, третья - c2\*o_x, четвертая - c3\*o_x

Завихренность по бокам от полосы замедления оказываются сформированы за счет слагаемого c1. Завихренность в центральной части расчетной области, как и ожидалось, сформирована за счет наклона уже существующей завихренности, выражаемого слагаемым c3.

Из четырех слагаемых, составляющих c1, два, связанные с производной стационарного течения по х, точно равны 0:

$$
- u \frac{\partial \Omega_x}{\partial x} = 0
$$

$$
\omega_x \frac{\partial U}{\partial x} = 0
$$

Слагаемое $- (U - c_f) \frac{\partial \omega_x}{\partial x}$, хотя и отличается от 0, не может дать вклада в энергию продольной завихренности

$$
- \frac{ 1}{L}\int_0^{L} (U - c_f) \frac{\partial \omega_x}{\partial x} \omega_x dx = - \frac{U - c_f}{L}\int_0^{L} \frac{\partial}{\partial x} \frac{\omega_x^2}{2} dx = - \frac{U - c_f}{L} \left[ \frac{\omega_x^2}{2}\right] _0^L = 0
$$

Остается только одно слагаемое, которое отвечает за производство продольной завихренности в пульсационном движения по бокам от полосы замедления, и это 

$$
\Omega_x \frac{\partial u}{\partial x}
$$

Физический смысл этого слагаемого просто. Пульсационная продольная скорость смещает стационарную завихренность вдоль оси х. При этом среднее ее значение не меняется, но локально где-то ее становится больше, где-то меньше. Таким образом возникает пульсационная составляющая завихренности. 

Получается так, что не просто пульсации продольной скорости в фазе с пульсациями продольной завихренности, что ведет к образования стационарных продольных вихрей. Получается, что пульсации продольной завихренности тоже формируются за счет пульсаций продольной скорости, что объясняет их синхронизацию. Пульсации продольной скорости причина всему, и действуют как насос: если продольная завихренность есть, то ее амплитуда увеличивается. 

Получается замкнутый круг. Стационарная завихренность производит пульсационную, а пульсационная - стационарную. Возможно мы нашли механизм, через который завихренность просто перетекает из одного состояния в другое, а ее количество при этом не меняется? За ее производство отвечает какой-то более тонкий механизм, который мы пока игнорировали? Вроде бы ответ нет, так как каждое из выделенных слагаемых дает прирост энергии завихренности. 

## Замечание

С этого момента пульсационные составляющие движения отмечаются штрихом, а переменные без штриха соответствуют сумме пульсационной и стационарной части движения. Большие буквы по прежнему отмечают среднее течение. 

Получается интересная вещь, что пульсационная составляющая продольной завихренности формируется за счета стационарной через слагаемые (я сознательно оставил второе слагаемое, про которое только что говорил, что им можно пренебречь)

$$
\frac{\partial \omega_x'}{\partial t} = - (U-c_f)\frac{\partial \omega_x'}{\partial x} + \Omega_x \frac{\partial u'}{\partial x} + ...
$$

А стационарная составляющая продольной завихренности формируется за счет пульсационной через такие же слагаемые 

$$
\frac{\partial \Omega_x}{\partial t} = - u' \frac{\partial \omega_x'}{\partial x} + \omega_x' \frac{\partial u'}{\partial x} + ...
$$

Эти два уравнения можно сложить. Учитывая, что 

$$
\frac{\partial u'}{\partial x} = \frac{\partial u}{\partial x}
$$

$$
\frac{\partial \omega_x'}{\partial x} = \frac{\partial \omega_x}{\partial x}
$$

получаем

$$
\frac{\partial \omega_x}{\partial t} = - u \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x} + ...
$$


Таким образом производство суммарной продольной завихренности выражается только через суммарные величины. Можно ли обобщить результат, сказав, что любые пульсации - периодическое сжатие и растяжение жидкой частицы приводит к тому, что завихренность вдоль направления сжатия возрастает? Какой за этим кроется физический механизм?

Выводы о том, что и в полном течении (без разделения на среднюю и пульсационную части) за производство продольной завихренности отвечают только два нелинейных слагаемых, можно проверить. Уравнение на продольную завихренность 

$$
\frac{\partial \omega_x}{\partial t} = - (u - c_f) \frac{\partial \omega_x }{\partial x}  - v \frac{\partial \omega_x }{\partial y} - w \frac{\partial \omega_x }{\partial z} + \omega_x \frac{\partial u}{\partial x} + \omega_y \frac{\partial u}{\partial y} + \omega_z \frac{\partial u}{\partial z} + \nu \Delta \omega_x
$$

Как и раньше, нелинейные слагаемые разделим на три группы

Конвекция и растяжение вдоль трубы
$$
сux + dxx = - (u-c_f) \frac{\partial \omega_x }{\partial x} + \omega_x \frac{\partial u}{\partial x}
$$

Конвекция в поперечной плоскости
$$
cvx+cwx = - v \frac{\partial \omega_x }{\partial y} - w \frac{\partial \omega_x }{\partial z} 
$$

Поворот существующих вихревых линий
$$
dyx + dzx = \omega_y \frac{\partial u}{\partial y} + \omega_z \frac{\partial u}{\partial z}
$$

## Заключение

Исследуя нижнюю ветвь бегущей волны в канале удалось получить замечательные результаты. 

Я считаю, что они были получены в первую очередь благодаря тому, что амплитуда пульсаций на нижней ветви небольшая. Это привело к тому, что среднее течение в ней воспроизводит полосу замедления достаточно точно, чтобы воспроизвести пульсационную составляющую движения в рамках линеаризованных уравнений. Нелинейным взаимодействием пульсаций можно пренебрегать. Их роль остается только в том, чтобы ограничить рост пульсационного движения. Более того, небольшая амплитуда пульсации делает возможным наш подход к исследованию механизма самоподдержания. Когда исследуешь бегущую волну с большой амплитудой пульсаций, среднее течение уже не точно повторяет полосу замедления, и теряет устойчивость к возмущениям другой формы. Нелинейное взаимодействие пульсаций существенно, им нельзя пренебречь. Кроме того, можно посчитать, что при исследовании механизма самоподдержания, мы неявно выделяем линии, на которые проектируем скорость и завихренность, и вдоль которых осредняем. В нашем случае они параллельны оси х. Если амплитуда пульсаций не велика, то они параллельны полосе замедления, и наш подход работает. Если полоса замедления изгибается сильно, то, вероятно, и линии нужно брать кривые, повторяющие форму полосы замедления, чтобы получить те же результаты. 

Неустойчивость, возникающую на полосе замедления, действительно можно схематизировать, как последовательность нормальных к стенке вихрей, расположенных на полосе замедления, чередующихся по знаку. Они дают du'/dx в области, где формируются стационарные продольные вихри. В этой же области наблюдается ox' нужного знака. 

Основной вопрос, который возник у нас при исследовании механизма самоподдержания - определить, что приводит к возникновению ox'. Оказывается, наклон вертикальных вихрей на фоне градиента средней скорости действительно имеет место, но он создает продольную завихренность в центре расчетной области по z выше полосы замедления. В области, где формируются стационарные продольные вихри ox' формируется за счет другого механизма, а именно за счет слагаемого 

$$
\Omega_x \frac{\partial u'}{\partial x}
$$

Это объясняет синхронизацию ox' и u'

Получается так, что в присутствии du'/dx стационарная OX производит пульсационную ox' , а пульсационная ox' производит стационарную OX. Как это может быть, это же какой-то замкнутый цикл? 

##  Appendix A

В рассуждениях выше я предполагал, что механизм образования стационарных продольных вихрей здесь тот же, что и в решениях, которые мы смотрели раньше. Я предполагал, что стационарная продольная завихренность OX или $\Omega_x$ формируется за счет нелинейного взаимодействия пульсаций (u,v,w) через слагаемые 

$$
\frac{\partial \Omega_x}{\partial t} = - u \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x} + ...
$$

(В этом разделе маленькие буквы - пульсационная составляющая движения, большие - стационарная)

Стоит проверить, действительно ли и в случае этого решения стационарные продольные вихри формируются также. Вклад различных слагаемых в энергию OX приведен на следующем рисунке. 

<img src="OX_gen.png">

Вторя картинка - вклад пары слагаемых, приведенных в уравнении выше. Как и предполагалось, они ответственны за возникновение OX. Другие слагаемые перечисленных ниже. 

$$
cux + dxx = - u \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x}
$$

$$
cvx + cwx = - v \frac{\partial \omega_x}{\partial y} - w \frac{\partial \omega_x}{\partial x}
$$

$$
dyx + dzx =  \omega_y \frac{\partial u}{\partial y} + \omega_z \frac{\partial u}{\partial z}
$$

Большие буквы - те же слагаемые, но вычисленные от стационарной составляющей движения.
