# Сеточная сходимость

Нами было найдено решение на сепаратрисе при Re = 2200 на сетке

    Xmax = 120, epsr = 0.25
    Im * Jm * Km = 1024 x 40 x 32
    
Затем оно было продлено в пространстве параметров так, что с нижней ветви решения удалось перейти на верхнюю. Сейчас мы собираемся писать статью, в которой будут описаны результаты, полученные при исследовании решения на сепаратрисе и решения с верхней ветви. Для того, чтобы иметь возможность о них писать, необходимо продемонстрировать сеточную сходимость, то есть показать, что при дальнейшем увеличении числа узлов в сетке и увеличении её протяженности решение не меняется. 

Что значит, что решение не меняется? Для нас важно, чтобы оно не менялось качественно и не значительно меняется количественно при значительном увеличении числа узлов в сетке и размера расчетной области. Я думаю, что увеличение длины расчетной области вдвое и уменьшение шага сетки вдвое можно считать значительным. Тогда новое решение следует искать на сетке 

    Xmax  = 240, epsr = 0.25
    Im * Jm * Km = 4096 x 80 x 64
    
По-моему, это невозможно, да и не нужно. 

## Сетки, на которых был выполнен расчет

Оказалось, для того, чтобы адекватно воспроизвести решения, достаточно сетки меньших размеров. Решения, полученные на сетке, параметры которой приведены ниже, качественно не отличаются от решений, полученных на исходной сетке. 

    Xmax = 80, epsr = 0.25
    Im * Jm * Km = 512 x 32 x 32 
    
Продемонстрировать сеточную сходимость на такой сетке уже возможно. Для того, чтобы это сделать, было выполнено две серии расчетов. В первом случае шаг сетки по всем направлениям был уменьшен вдвое. Расчет на такой сетке позволил показать, что от шага сетки решение существенно не зависит. Параметры этой сетки: 

    Xmax = 80, epsr = 0.25
    Im * Jm * Km = 1024 x 64 x 64
    
Во втором случае была увеличена длина расчетной области, что позволило показать, что от нее также ничего не зависит. Сетка при этом была следующая: 

    Xmax = 160, epsr = 0.25
    Im * Jm * Km = 1024 x 32 x 32
    
Первая серия расчетов на наиболее легкой сетке заняла примерно 2 дня времени непрерывного счета (получить решение на сепаратрисе и перейти на верхнюю ветвь). Поиск решений на наиболее плотной сетке занял примерно неделю времени непрерывного счета. 

Я пробовал менять только шаг по пространству, не не трогал шаг по времени. Думаю, в этом нет необходимости, так как он выбирался во время счета программы автоматически из условия tol = 0.01. Хотя можно было бы показать, что и при уменьшении шага по времени вдвое решение также не меняется. 

В статье можно представить результаты, полученные на любой из сеток. Исходная сетка занимает промежуточное положение между сетками, на которых решение было пересчитано. На новой сетке длина расчетной области равна 80R, как в статье Авила 2013, где это решение было найдено. Как-то возникал вопрос, почему я выбрал Xmax = 120R в своих расчетах. Если использовать сетки Xmax = 80R, то такой вопрос не возникает. Кроме того, на новой сетке количество точек по радиусу -- степень двойки. Может быть, и это лучше, чем Jm = 40, как это было в исходном расчете. 

Я использую значение epsr = 0.25 основываясь по большому счету не на чем. Я пробовал оценить, чему могло бы быть равно epsr, изменяя его методом Ньютона на верхней ветви решения на исходной сетке. В диапазоне от 0.15 до 1.0 от его значения решение зависит достаточно слабо, и, в принципе, все решения могут быть получены при epsr = 1 без растяжения сетки вообще. Однако сейчас все уже посчитано при epsr = 0.25, и есть сеточная сходимость. 

## Сравнение решений на различных сетках

Возникает еще один вопрос, что сравнивать? По видимому, нужно сравнивать наиболее сложные величины, которые будут представлены в статье. Это, например, распределения в сечениях различных слагаемых, отвечающих за генерацию стационарных продольных вихрей и их пульсаций. Это результаты решения линеаризованных уравнений. Пока я их не смотрел, а здесь будут приведены сравнения полей скорости и параметров решения. 

На графике ниже изображено, как меняется период решения во времени T и скорость его сноса cf по мере продвижения в пространстве параметров (в зависимости от Re). Какая линия к какой сетке относится, конечно, не разобрать, но это и не важно. Важно, что они близки друг к другу на протяжении почти всей траектории. На нижней ветви значения T и cf оказываются выше, чем на верхней, поэтому нижняя ветвь решения на графиках оказывается выше. Траектории начинают существенно расходится на верхней ветви решения при Re > 1700 примерно.

На некоторых сетках я так и не смог продвинуться до Re = 2200 на верхней ветви. Напомню, что продвигаясь по верхней ветви решения на исходной сетке я столкнулся с тем, что не мог пройти точку Re ~ 1870. Мне удалось её обойти сместившись в сторону по nsym. По видимому, с этой же проблемой я столкнулся на самой подробной сетке, и пока довести решение до Re = 2200 на верхние ветке на этой сетке не смог. На более мелких сетках с этой проблемой я не сталкивался. 

Я думаю, нет смысла доводить решение на верхней ветви до Re = 2200, а стоит остановиться на Re = 1700, например. При этом Re все траектории еще близки друг к другу. Кроме того, при движении дальше возникают эти точки, которые я не могу пройти прямо, и не понимаю, в чем может быть причина такого поведения.

## Зключение

Думаю, полученные результаты достаточно ясно демонстрируют то, что нижняя ветвь решения и верхняя до Re = 1700 воспроизводятся в наших расчетах адекватно. Если мы хотим дотянуть до Re = 2200 на верхней ветви, то нужно пересчитывать все на более подробной сетке, что потребует очень много времени. Поэтому, я думаю, на верхней ветви стоит остановиться при Re = 1700. 

Есть еще возможность двинутся в сторону увеличения Re на нижней ветви. Так можно еще немного увеличить разницу между решениями, пригодными для равнения в статье. Также можно показать, что с ростом Re на нижней ветви решение приближается к оси трубы. Правда я затрудняюсь сказать, какие из этого следуют выводы.
