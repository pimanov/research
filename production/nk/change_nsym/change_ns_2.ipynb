## Причины отсутствия сходимости метода Ньютона-Крылова

Причины, по которым решение перестает сходиться, можно разделить на две независимые. Метод Ньютона-Крылова включает в себя два итерационных процесса, которые сходятся/расходятся независимо друг от друга. Первый состоит в отыскании градиента функции F(x) в точке x0. Для этого решается линейная система большой размерности итерационным методом, основанным на подпространствах Крылова. Второй относится к методу Ньютона - в соответствии с найденным значением градиента в точке x0 находится новое приближение к решений x1. 

Если решения не существует, то метод Ньютона конечно не сойдется, но градиент все равно существует, и Крыловский метод должен сходиться. В то же время, метод Ньютона может расходиться и просто, если начальное приближение "слишком далеко" от решения. 

Как правело, уже найденное решение становится начальным приближением при соседних значениях параметров. При этом всегда при достаточно маленьком размере шага по параметрам от найденного решения к новому, найденное решение будет хорошим начальным приближением к новому решению и метод Ньютона сойдется. Следовательно, если метод Ньютона не сходится, можно попробовать уменьшать шаг. 

Если решения не существует, граница, за которой его не существует и метод Ньютона перестает сходиться, должна четко вырисовываться. Например, если решение при Re=1000 не сходится, когда я пришел к нему от Re=999 с шагом 1, а с шагом 0.6 я смог найти решение и при Re=999.6 и при Re=1000.2, то границы здесь нет, просто шаг 1 был слишком большим, чтобы двигаться к новым решениям. 

Таким образом, если не сходится метод Ньютона, то уменьшая шаг по параметрам, можно либо вернуть ему сходимость, либо сказать, что решения не существует. 

Если не сходится Крыловский метод, то о решении ничего уверенно сказать нельзя. Нужно искать способы повысить сходимость метода. 

Если еще много параметров, значение которых влияет на результат. Например, какого размера выбирать шаг при численном вычислении производной, при вычислении градиента? С какой точностью искать решения крыловским методом? Как влияет точность, с которой было найдено предыдущее решение на то, будет ли оно хорошим начальным приближением при поиске нового решения? Их влияние сложно оценить, и поэтому утверждать, существует ли решение или нет можно лишь с определенными оговорками. 

## Однозначность определения решения по заданным параметрам

Стоило бы еще проверить, сохраняется ли на верхней ветви свойство, что по двум параметрам из четырех (T,cf,Re,nsym) можно однозначно найти решение. И показать, что это было так на нижней ветви. 

Я вижу только одни способ это аргументировать. Показать, что если оставлять на самоопределение только один любой параметр, то в направлении градиента будет лишь небольшое уменьшение функции F(x), в то время, как если на самоопределение отдать любые два, то в направлении градиента F(x) приблизится к нулю на порядок/два, если начальное приближение достаточно близко к решению. При этом остается много вопросов. То, что в направлении градиента функция F(x) не приближается к нулю можно интерпретировать просто так, что начальное приближении "слишком далеко" от решения. Или это можно интерпретировать так, что значение градиента просто вычислено недостаточно точно. Могут быть и другие причины, так что доказать так что-то сложно.
