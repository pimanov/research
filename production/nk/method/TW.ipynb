Возможно более удобным обектом для исследования формирования продольной завихрености являются бегущие волны TW. Глядя на иллюстрации, приведенные ниже, создается впечатление, что TW устроены так же, как и ядро нашего локализованного решения. TW представляют собой бесконечно длинные полосы, однородные вдоль трубы, по которым бежит та же волна, что и в ядре нашего решения, поодерживающая их существование. Соответственно скорость сноса порыва пропадает, остается только скорость бегущей по порыву волны. Можно перейти в систему отсчета, связанную с волной, и в ней все решение стационарное.

$$

Изображены несколько сечений трубы, если мы считаем, что находимся в сопуствующей системе отсчета. Если считать, что мы в системе отсчета, связанной с трубой,то это несколько последовательых моментов времени. Здесь отчетливо видны полосы повышеной и пониженой скорости. Как и у нас, вихри возникают поочереди то с одной, то с другой стороны около полосы замедления, между полосами замедления и ускорения. Как и в нашем случае, плоскость, относительно которой наблюдается симметрия отражения со сдвигом, проходит через полосу замедления. Главное отличие в том, что полосы ускорения тут идут парочками, и между ними похоже нет взаимодействия, нет пульсаций.

Иллюстрация из работы <a href="Wedin2004.pdf">Wedin2004.png</a> "Exact Coherent Structures in Pipe Flow: Travelling Wave Solutions". В этой работе найдено три семейства бегущих волн. Каждое семейство здесь представлено двумя изображениями, расположенными в одной строке. Левая - верхняя ветвь решения, правая - нижняя ветвь. В данном случае верхней ветвью называют ту,в которой скорость сноса выше, соответственно в нашем случае это наоборот нижняя ветвь. Среди них есть не только такие, как на предидущей иллюстрации, где полосы ускорения встречаются по две подряд, но и такие, как в нашем случае, где полосы ускорения и полосы замедления чережуются одна через одну. 

Здесь приведены только осредненые вдоль трубы решения, но из других иллюстраций в этой работе можно догадаться, что в них также вихри возникаю поочереди, то с одной, то с другой стороны от полосы замедления. 

Бегущие волны, найденые в приведенных работеах, объединяет то, что пульсации возникают вокруг полос замедления, причем с двух сторон, поочереди. Полосе ускорения же сполне достаточно, чтобы пульсации происходили только с одной стороны от нее. В третьем семействе решений на втором рисунке возникают полосы замедления, вокруг которых вообще не происходит пульсций - это, вероятно, просто пониженая скорость из центра проникает в область между двумя полосами ускорения, так как между ними есть место. 

Может ли так быть, что пульсации возникают только в одном промежутке между полосой ускорения и змедления, и поддерживает их существование? Может ли так быть, что пульсации образуются вокруг полосы ускорения с двух сторон, а полосы замедления рядом будут поддерживаться пульсацимя толкьо с одной стороны?

Когда по полосам бугет пульсации, они формируют продольную завихреность, которая подеерживает существоавние полос. Важную роль в данном процессе играет вязкость. С одной стороны, вязкость ограничиват значение продольной компоненты завихерности. С другой стороны, в присутствии постоянного значения продольной завихерности, полосы тоже росли бы неограничено, если бы не вязкость. 

Можно попробовать сделать следующею вещь. Разложить поле скорости бегущей волны на среднюю и пульсационную составляющей (осреднить вдоль трубы). Затем для средней составляющей отключить вязкость и удалить поперечное движение. Такое поле скорости является решением уравнений Эйлера и продолжило бы сущетсовать с потоке неограничено долго. Более того, завихреность в нем не имеет продольной компоненты, и вихревые линии перемещаются вместе с частицами жадкости. 

Теперь можно пойти двумя путями. Можно посмотреть, как будет эволюционировать данное поле скорости в присутствии пульсационной составляющей движения. Возмоно удастся наглядно увидеть, как пульсации деформируют вихревые линии основного течения так, что в нем возникает и продольная компоненты завихерности. 

Можно просто посчитаь эволюцию решения с выключеной вязкостью. Так как поле скорости неустойчиво, то пульсации на нем возникунут сами собой, и можно будет наблюдать за деформицией вихревых линий. Отличие от первого способа в том, что вихрвевые линии вуду посттроены и по пульсационной составляющей скорости тоже, и будут совершать некоторые циклические движения в пространстве, среди которых нужно будет суметь распознать, как меняется продольная компоненты завихерности в среднем. 

Так как вихревые линии в идеальной жидкости перемещаются вместе с частицмаи жидкости, можно следить именно за перемещением вторых, что знать, как менятся первые. Дл этого можно ввести некоторую инертную примесь, которая перемещается в потоке, и расчитовать ее перемещение параллельно основной программе. 

В первый момент времени задаем два поля примели. В одном значения совпадают со значением х. В другом - со значением скорости. В таком случае полосы, на которых пурвое и второе поле скорости - константы - в точности совпадают с вихревыми линиями. Тоже сохранится и с течением времени.
