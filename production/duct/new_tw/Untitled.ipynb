# Замечания

На рисунке ниже приведено поле скорости, соответствующее бегущей волне в канале. Продольные сечение равноудалены друг от друга и заметают один период волны. Первое сечение повторяет последнее. Изображена продольная скорость u. 

<img src="umap.png">

Полоса замедления ярко выражена, узкая и вытянутая. Она отклоняется то влево, то в право (синусоидальная неустойчивость), при этом смещается больше, чем на собственную ширину. Это приводит к тому, что после осреднения полоса сильно размывается и оказывается значитльно более устойчива. И если на среднем течении и будут нарастать какие-то пульсации, их уже нельзя прямо связать с наблюдаемыми. Хотя исследуя и их можно, вероятно, получить интересующие нас представления, так как они достаточно универсальны, по-видимому. 

Можно было бы исследовать локализованное решение, так как в нем есть область в которой пульсации на полосе замедления только возникают и находятся на линейной стадии развития. Я решил проверить, можно ли также, как я получил эту бегущую волну, получить локализованное решение, если увеличить лину расчетной области. У меня не вышло. В таком случае решение при Re=1400 даже не существует, оно быстро ламинаризуется. 

Если вернуться к бегущей волне, вместо того, чтобы исследовать на линейную устойчивость среднее течение, стоит вбрать такое решение, которое предствляет из себя именно такую полосу замедления, но не изогнутую, симметричную относительно вертикального сечения, постоянную вдоль канала. Такое поле скорости вседа можно сконструировать искуственно, например выбрать то сечение, в котором полоса направленна вертикально вверх, и размножить его на всю трубу. Можно ли выделить такое течение как-то иначе, более строго? Как быть с течениями в поперечной плоскости, нужно ли их сохранять? Влияют ли они на что-то - это можно проверить просто. 

Какой можелью можно приблизить эту полосу замедления, чтобы получить форму. Какая особенность этих полос определяет форму пульсаци и их наличие? 

То, что полосы так вытянуты, повидимому не должно иметь большого значения, так как и для решения на сепаратрисе механихм поддержания такой-же несмотря на то. что полосы там едва заметны. 

Какую роль имеет давление? можно ли его игнорировать? По крайней мере непериодическую составляющюю давления можно игнорировать, так как его эффект только в том, чтобы проталкивать жидкость вдоль. 

Я ожидал, что размер полосы будет уменьшаться с увеличеснием Re, но это оказалось не та. Даже при Re=3000 полоса занимает все туже часть трубы, и имеет лишь больше искажений и других особенностей. 

Как вообе течение, в котором есть только продольная скоорость, может получить скорость и в других направлениях, в частнсти в направлении от стенки. 

Стоит ли рассматривать вихри? Достаточно, чтобы поток имел вертикальную скорость, горизонтальная же возникнит в нашем случае уже просто из условия непротикания на стенках. 

Что может дать вертикальную скорость? 

Почему нижняя часть полосы не движется, а раскачивается только ее верхняя часть?

## Второе

Моно исследовать уравнения, выделять среднее и пульсационые составляющие и т.п. Но сердце требует объяснения на языке пальцах, на языке жидких масс и их инерции, на материальном языке. Такое объяснение может быть лучше потому, что дает естественное ощющение конвекции - единственного механизма, которые все это формирует. С другой стороны смотреть за каждой частицей индивидуально нет смысла, так как они все одинаковы, и заменяют друг друга без последствий. Непонятно, можно ли получить линейную неустойчивость на материальном языке.
