# Бегущая волна в пространстве параметров

Сейчас бегущая волна занимает практически все сечение канала, как по высоте, так и по ширине (рис.1). При этом стенки оказывают значительное влияние на форму течения. Являются ли они существенными для самоподдержания этого решения или только вносят искажения в такой процесс, который мог бы существовать и при удалении стенок на бесконечность? Может так быть, что в результате неустойчивости пульсации имеют только продольную компоненту скорости, а вихри образуются просто как следствие того, что жидкость упирается в стенку. В такой модели стенка необходима. Можно было бы попробовать отодвинуть стенки, увеличить расчетную область, и посмотреть, как при этом изменится решение. 

<img src="umap.png"> Рис. 1. Исходная бегущая волна в канале. Продольная компонента скорости. Цветовой масштаб совпадает на этом и всех рисунках ниже (синий - 0, красный - 1). Один период волны, первое сечение повторяет последнее. Параметры бегущей волны:

    Xmax * Ymax * Zmax = 5 x 1 x 1
    Im * Jm * Km = 64 x 40 x 32
    Re = 1400
    cf ~ 0.73 

Первое, что приходит в голову - увеличить Ре. Это равносильно тому, чтобы увеличить расстояние Ymax. Так как Xmax и Zmax задаются через Ymax, увеличение Ре равносильно увеличению размера расчетной области с сохранением Ре и пропорций между сторонами расчетной области. Может быть так, что полоса замедления будет уменьшаться относительно размеров расчетной области, оставаясь в центре. Это значило быть, что стенки не существенны для решения, оно локализовано и в поперечном направлении. Может быть так, что решение окажется зацепленным за стенки (верхнюю или боковые), тогда они являются существенной частью этого процесса.

## Зависимость от Zmax

Последний параметр, который остается - это Zmax. Меняя его возможно удастся оторваться от боковых стенок. 

<img src="cf_Zmax.png" style="float: left"> Re=2200. Решение существует при 0.9 < Zmax < 1.6. При больших и меньших Zmax получить решение у меня не получилось, хотя я не могу гарантировать, что его там нет. Скорость сноса опять практически не меняется. 

Ниже распределение поля скорости при Zmax = 1.06, 1, 0.9. 

<img src="maps_varZmax.png">

Похоже, что ширина канала определяет размер бегущей волны, что боковые стенки имеют непосредственное отношение к поддержанию такой волны. При изменении Ре размер волны меняется ни чуть не меньше, но там я говорю, что от Ре решение не зависит. Дело в том, что там Ре меняется почти в 5 раз, а здесь Zmax меняется только на 16%. Примерно на столько же меняется и размер полосы замедления. Под размером полосы замедления я имею ввиду только ее высоту и ширину.

## Заключение

Можно выделить две тенденции. От Ре форма полосы замедления практически не зависит, то есть вязкость не существенна в процессе самоподдержания решения. Форма бегущей волны в поперечном сечении постоянна, ее размер определяется Zmax - периодом в трансперсональном направлении. Зависимость от Xmax, то есть длины волны в продольном направлении, неочевидна. 

У меня остаются сомнения, можно ли так интерпретировать полученные результаты. Метод Ньютона продолжает решение непрерывным образом, при этом должны сохраниться не только существенные для поддержания волны особенности, но и просто все. Следствие этого я считаю, например, появление "бородавок" на последнем рисунке внизу, как во многих других случаях. Среди особенностей могут быть и такие, которые связывают решение с боковыми стенками, и даже если они не существенны, метод Ньютона их сохранит. Возможно, не совсем ясно, что я тут хочу сказать, но причина этого в том, что мне самому не ясно. 

Кроме метода Ньютона для поиска решений при новых значениях параметров можно использовать тот же метод, каким было получено исходное решение - выделить бегущую волну из турбулентного течения. В этом случае сделанные выводы уже имеют непосредственное отношение к турбулентности, все несущественные особенности определяются непосредственно существенными, и не влияют на них. Можно ожидать, что продолжения, полученные методом Ньютона и решения, полученные непосредственно из турбулентного течения, в некотором диапазоне параметров совпадут. Это оправдало бы сделанные выводы в этом диапазоне параметров. 

Даже если сделанные выводы не справедливы, данные о том, как зависит решение от параметров, останутся актуальны и верны, так что этот файл имеет ценность.
