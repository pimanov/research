# Механизм самоподдержания бегущей волны. Набросок.

Бегущая волна все та же (рис.1). У меня наконец-то созрела некоторая идея, как объяснить причину того, что $\partial u / \partial x$ и $\omega_x$ совпадают по фазе, что ведет к формированию продольных вихрей. Идея очень простая.

## Общая идея

Можно представить механизм самоподдержания бегущей волны следующим образом. 

1. Частицы жидкости около стенки получают вертикальную составляющую скорости и поднимаются над стенкой, образуя тем самым полосу замедления.

2. Полоса замедления оказывается неустойчива, по мере подъема частицы жидкости начинают раскачиваться, то влево, то вправо. По полосе замедления бежит волна. Механизм схож с образованием волн на поверхности воды. 

Это все понятно, важен пункт 3, посвященный тому, как формируются продольные вихри.

3 . Когда, допустим, волна замедления смещается влево, она толкает жидкость вне полосы замедления в эту же сторону (рис.2). Жидкость растекается на стенке как вверх, так и вниз, и растекаясь вниз создает продольную завихренность $\omega_x$ нужного знака в нужном месте - сбоку от полосы замедления. Одновременно с этим, смещаясь в сторону стенки, полоса замедления выдавливает жидкость из этой области, двумерная дивергенция $\partial v / \partial r + \partial w / \partial z < 0$, соответственно возникает $\partial u / \partial x > 0$. Получается, что $\omega_x$ и $\partial u / \partial x$ возникают в результате действия одной и то же причины, и они синхронизованы, как друг с другом, так и с фазой волны. - Это ключевая идея.

<img src="sceam.png" style="float: left"> Рис.2. Схема того, как бегущая волна, сдвигаясь влево, создает продольную завихренность и одновременно производную продольной скорости нужного знака. 

Если считать, что $\omega_x$ и $\partial u / \partial x$ меняются вдоль трубы, как синус/косинус, так как возникают в результате линейной неустойчивости, то и $u$ и $\omega_x / \partial x$ тоже меняются вдоль трубы, как синус/косинус, и тоже синхронизованы друг с другом. Таким образом все слагаемые, которые дают вклад в продольные вихри, имеют нужную фазу 

$$
- u \frac{\partial \omega _x}{\partial x} + \omega_x \frac{\partial u}{\partial x}
$$

В данной схеме стенка принимает самое непосредственное участие. 

Эта схема пока взята просто из головы, и нужно еще выяснить, какое отношение она имеет к тому, что происходит в бегущей воле в действительности.

## Связь с действительностью

Если посмотреть на поле завихренности (рис.3), то окажется, что оно распределено немного иначе. Как и раньше, на этом рисунке двигаться вперед по времени, значит двигаться назад по фазе. Если начать с картинки 5pi/16 и двигаться в сторону уменьшения фазы, то видно, что полоса замедления еще не начала двигаться влево, но положительная завихренность уже формируется. И формируется не только в близи стенки на высоте волны, но во всем пространстве. Это несоответствие можно объяснить так, что сейчас механизм уже работает, и полоса замедления вся состоит из частиц жидкости, двигающихся вверх. Они как фонтан бьют из вершины полосы замедления вверх, постоянно создавая завихренность нужного знака по бокам от полосы замедления. 

<img src="om.png"> Рис.3. Цвет - продольная завихренность. Линии - изолинии продольно скорости. Поле скорости бегущей волны, Ре=1400. Один период вдоль трубы, первое сечение повторяет последнее. 

Но есть еще одно обстоятельство. Для того, чтобы механизм работал, необходимо, чтобы жидкость, окружающая полосу замедления, двигалась быстрее волны. Чтобы в системе отсчета жидкости волна бежала по полосе замедления, а не стояла на месте. Это в данном случае крайне плохо выполняется, как показано на рис. 4. 

 <img src="cf.png"> Рис.4. Цвет - продольная скорость. Линия отделяет жидкость, которая движется быстрее волны от жидкости, отстающей от нее. 

Там, где полоса замедления прижимается к стенке, жидкости, двигающейся быстрее волны уже практически не остается, и ее уж точно не достаточно, чтобы поддержать 
тот процесс. Только в сечениях 14pi/8 и 6pi/8 можно найти небольшое количество жидкости, обгоняющей бегущую волну, зажатый между полосой замедления и стенкой. 

Но на этом же рисунке можно найти достаточно правдоподобное объяснение того, что тут происходит. В тот момент, когда полоса замедления наклоняется вправо (сечение 6pi/8), поток жидкости, поднимающийся из вершины полосы замедления, начинает наполнять правый верхний угол, это видно на сечениях 6pi/8 - 3pi/8. Эта жидкость точно также, как и в исходной модели, создает продольную завихренность и теснит жидкость, которая там была до этого, создавая производную продольной скорости вдоль трубы, но уже в пространстве над волной, в верхней части трубы. И в этой области образуются и развиваются продольные вихри, которые потом смещаются под полосу замедления, и там только затухают.

Теперь для процесса поддержания решения оказываются важны не только боковые стенки, но и верхняя. Это уже совсем не похоже на то, что могло бы происходить в реальной турбулентности. 

## Заключение

Есть базовая идея, почему продольная завихренность и производная вдоль трубы от продольной скорости оказываются в фазе, и как они возникают. Эта идея тут возникает в разных вариациях, и наверное применима и для турбулентности, но в каком-то третьем виде, так как в данном случае процесс непосредственно связан с наличием стенок. 

Второй вариант объяснения механизма образования вихрей мне кажется достаточно правдоподобным, но есть еще много вопросов к нем. Например, может так быть, что колебания полосы замедления связанны с присутствие больших вихрей в верхней части канала, которые смещают ее то влево то вправо, и они являются основной силой здесь. Нужно привести еще аккуратно все обосновать перед тем, как что-то утверждать, это только первый набросок. 

Было бы интересно объяснить с точки зрения механики, а не математики, почему слагаемые типа $\omega_x \partial u / \partial x$ ведут к образованию продольных вихрей. Возможно это объяснение можно лаконично включить в общую картину. 

Можно попробовать посчитать линейную устойчивость полосы замедления, расположенной вдали от стенок, чтобы как-то приблизиться к описанию реальной турбулентности. Найти форму возмущений и области формирования продольной завихренности без стенок.
