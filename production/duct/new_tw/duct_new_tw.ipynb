# Новая бегущая волна в канале

Оказывается в канале, как и в трубе, можно найти турбулентный режим, очень близкий к бегущей волне. Подав такое течение на вход методу Ньютона-Крылова удается ее выделить в чистом виде, что я и сделал. С этим решением уже можно работать в новой геометрии прямоугольной трубы. 

## Турбулентное течение

Чтобы получить такое решение, я просто уменьшал число Рейнольдса и смотрел, как меняется турбулентное течение. При Re=1400 решение стало похоже на бегущую волну. На графике ниже изображена зависимость амплитуды пульсаций от времени.  При t чуть меньшем, чем 2000, я уменьшил Re до 1400, и график вышел примерно на постоянное значение, что соответствует бегущей волне. Это решение оказывается очень устойчивым, и может существовать в таком виде тысячи единиц по времени, если не бесконечно долго. 

<img src="new_tw_prt.png">

Такое решение продолжает существовать и при меньших числах Рейнольдса. На графике ниже приведена амплитуда пульсаций в зависимости от времени. Раз в 1000 единиц по времени я снижал Re скачком на 100 единиц. При Re=1300, 1200, 1100 решение продолжает существовать, при Re=1000 решение через некоторое время ламинаризуется. 

<img src="prt1.png">

Существуют ли другие турбулентные решения в этом диапазоне Re - я не знаю. Тут важно отметить, что расчет я начинал с решения на сепаратрисе, и не позволял развиться турбулентности так, чтобы она успела забыть начальные данные. При развитии турбулентности из решения на сепаратрисе вихри начинают перемещаться к стенке, набирают силу, происходят перестройки течения, но долгое время сохраняется симметричность течения относительно сечения $\theta = \pi/4$, если говорить на языке трубы. В данном случае я не дожидался, пока течение потеряет эту симметрию, когда начинал уменьшать Re, и всегда работал с симметричным течения. Возможно если бы я позволил турбулентному течению потерять симметрию, то с уменьшением Re я бы вышел на другое решение, не такое простое. 

## Бегущая волна

Подав турбулентное течение на вход в метод Ньютона-Крылова, удалось выделить из него бегущую волну. Параметры:

    Xmax * Ymax * Zmax = 5.0 x 1.0 x 1.0
    Im * Jm * Km = 64 x 40 x 32  
    Re = 1400  
    cf = 0.73

На графике ниже приведены средние (вдоль трубы): продольная скорость U, продольная завихренность OX, и пульсации puls. Еще указаны максимальные значения, которые достигает величина. 

<img src="means.png">

Даже при таком небольшом числе Рейнольдса пульсации не занимают все пространство, а локализованы около стенки. Полосы ярко выражены. Центры продольных вихрей и центры пульсаций достаточно сильно не совпадают (как и раньше). 

На графике средней скорости изолинии снова подходят под углом к верхней стенке. График ниже, я надеюсь, поможет прояснить ситуацию. На нем изображено значение U, как функция от координат (y,z). Около верхней стенки (y=1) не только нормальная производная близка к нулю, но и касательная тоже. Отсюда получается, что изолинии могут походить под значительным углом к ней.

## Заключение

Один и тот же метод позволил найти бегущую волну как для трубы, так и в канале. Возможно ,он обладает некоторой универсальностью и стоит написать об этом статью? Получается не просто бегущая волна, а такая, около которой и держится турбулениность, хотя и в таких редуцированных условиях (малый размер расчетной области и условия симметрии).

Теперь есть нормальная бегущая волна для исследования в ортогональной системе координат, можно с ней работать.
