## Средняя и пульсационная составляющие течения

Полоса замедления ярко выражена, узкая и вытянутая вверх. Она отклоняется то влево, то вправо (синусоидальная неустойчивость), при этом смещается больше, чем на собственную ширину. Это приводит к тому, что после осреднения полоса приобретает совершенно другую форму и оказывается значительно более устойчива (рис. 2). И если на среднем течении и будут нарастать какие-то пульсации, их уже нельзя прямо связать с наблюдаемыми. Исследовать взаимодействие средней и пульсационной составляющих течения в данном случае бессмысленно.

<img src="means.png" style="float: left"> Рис.2. Среднее поле скорости, завихренности и пульсаций, осредненные вдоль трубы.

В данном случае можно было бы исследовать на устойчивость поле скорости, представляющее собой такую же по ширине и высоте полосу замедления, но не изогнутую, расположенную в центре расчетной области. Видимо такое поле скорости может быть получено только искусственно. Например, можно выбрать одно из сечения, где полоса замедления направлена точно в вверх, и размножить его на всю трубу.  

Можно было бы исследовать локализованное решение, так как в нем есть область в которой пульсации на полосе замедления только возникают и находятся на линейной стадии развития. Я решил проверить, можно ли также, как я получил эту бегущую волну, получить локализованное решение, если увеличить длину расчетной области. У меня не вышло. В таком случае решение при Re=1400 даже не существует, оно быстро ламинаризуется. 

То, что полосы так вытянуты, по-видимому не должно иметь большого значения, так как и для решения на сепаратрисе механизм поддержания такой-же несмотря на то, что полосы там едва заметны. 

## Полоса замедления, как упругое тело

Полоса замедления ярко выраженя. В отличии от решения на сепаратрисе, она узкая и вытянутая в направлении от стенки. Можно отметить, что распределение скорости внутри полосы замеделния, ее форма и длина практически не меняются, она лишь изгибается то в лево то в право, как упругое тело. Возможно это только визуально так выглядит, но вероятно полоса замедления действительно состоит из одних и тех же частиц жидкости. Если бы с одной стороны в полосу замедления попадали бы новые частицы, то это тоже самое, что они бы резко замедлялись, в то время как с другой стороны от полосы частицы должны резко ускоряться, чтобы выйти из нее. Но я не вижу механизма, способно резко замедлять или ускорить частицы жидкости в продольном направлении. 

В качестве подтверждения, но не доказательства, что это так, хочу привести следующий рисунок. Не нем изображена серия сечений, параллельных стенке. Расстояние от стенки указано под сечением. Цветом показана продольная компонента скорости. Линиями - трансверсальная (theta), сплошные линии - положительное значение (скорость направлена вправо), прерывистые - отрицательное (влево). Двигаться вперед по времени - тоже самое, что двигаться по x в сторону убывания (движение сверху вниз в данном случае). 

<img src="xth_section_uw.png">

Если обратить внимание на полосу замедления и на то, какую трансверсальную скорость имеют частицы жидкости, из которых она состоит, то можно обнаружить подтверждение того, что полоса замедления и окружающее пространство не обменивается жидкостью. Там, где частицы жидкости смещаются вправо, трансверсальная скорость имеет положительное значение, там где вправо - отрицательное. Можно отметить, что вне полосы замедления всегда существует второй максимум трансверсальной скорости. С чем он может быть связан?

Возможно получить поле скорости вокруг полосы замедления можно, рассматривая ее просто как упругое тело с условие непротекания на поверхности, которое совершает колебательные движения. Такое представление поможет схематизировать процесс.

## Аналогия с волнами на воде

Тут можно вспомнить простой пример с волнами на поверхности воды. Математически есть вихревой слой, с разных сторон от которого скорость частиц отличается. При волнах на воде отличается еще и плотность, но это уже не существенно в данном случае. Такое течение всегда не устойчиво, и на пальцах механизм его неустойчивости можно представить следующим образом. 

Представим, что жидкость покоится, а ветер имеет скорость. Жидкость и воздух разделяет вихревой слой. В равновесном состоянии вихревые линии распределены равномерно по поверхности, они компенсируют друг друга и не вызывают никакого движения жидкости в поперечном направлении. Если теперь допустить неравномерность, то в той области, в которой сосредоточено больше вихревых линий, они вызывают движение поверхности воды. За этой областью жидкость поднимается, перед ней опускается. При этом поднявшаяся жидкость подгоняется ветром вперед, длина области повышенной завихренности сокращается, что приводит к фокусировке завихренности и развитию неоднородности. 

Это известное объяснение, его я слышал на лекция по теории устойчивости, но может сейчас что-то переврал, так как восстанавливал из головы. 

Этот механизм объясняет в частности, почему возникающая волна имеет скорость среднюю между скоростью жидкости над поверхностью и под ней. А если слой смешения размыт и скорость на нем меняется непрерывно, то скорость воны совпадает со скорость точки, в которой она возникает. 

Можно предположить, что в нашем случае механизм неустойчивости тот же. В таком случае полосу ускорения должны окружать вихревые линии. Они действительно есть. На рисунке ниже изображена нормальная к стенке завихренность.

## Вертикальная компонента скорости

Еще один интересный факт, что внутри полосы замедления компонента скорость v всегда отрицательна (движение от стенки). Область отрицательной v достаточно точно совпадает со полосой замедления. 

На рисунке ниже на этот раз линиями показана нормальная к стенке скорость v. Прерывистые линии - отрицательные значение. Вне области отрицательных значений значения положительны, линии в этой области не нарисованы. 

<img src="xth_sect_uv.png">

Интересно, что минимальное значение v имеет там, где полоса замедления сильнее всего отклоняется от положения равновесия, или может быть немножко раньше. Почему?  

На рисунке ниже изображено тоже самое, но в другое проекции. 

<img src="uvmap.png">

Почему область отрицательных v так точно совпадает с полосой замедления? Скорость v вполне могла бы быть местами и положительной, главное только чтобы в сумме движение оставалось движением от стенки. 

Может ли так быть, что наличие вертикальной скорости первостепенно и определяет положение полосы замедления? Если частица жидкости движется вверх, то становится частью полосы замедления, так как несет с собой меньшую продольную скорость. Если она останавливается в этом своем движений, то вязкость сразу же сравнивает ее скорость с окружающими частицами, и она выходит из полосы замедления. Эта точка зрения противоположна тому, что было написано выше, и мне она не кажется особенно правдоподобной. 

Другое объяснение может состоять в том, например, что частицы жидкости, получившие вертикальный импульс, сохраняют его при движении вверх. Тогда все частицы жидкости, составляющие полосу замедления, состоят из частиц, принесенных от стенки, имеющих отрицательную v. Но это не само собой разумеется, что те нелинейные процессы, которые создают вертикальное движение, потом ни разу на него не повлияют, что они только поддерживают это движение, но никогда не портят его.

В общем, объяснения у меня нет, это просто интересный факт. 

## Заключение

Я хотел отметить три факт, которые пока обнаружил. Они еще ничего не объясняют, но уже интересные. 

Во-первых, подход с разделением поля скорости на среднюю и пульсационную составляющие здесь неэффективен. Эффективно, мне кажется, смотреть за жидкими частицами. Для них нелинейное взаимодействие - это просто инерция, которую легко понять и представить себе, а кроме нее здесь ничего больше нет. Также эффективно стараться схематизировать течение. 

Во-вторых, стоит проверить, не обмениваются ли жидкостью полоса замедления и окружающее пространство. Если нет, то можно дать объяснение механизму возникновения пульсаций, и дать удобную схематизацию течения. 

В-третьих, частицы, которые составляют полосу замедления, всегда имеют компоненту скорости, направленную от стенки. 

В общем, это только начало. Стоит продолжать копаться в этом поле скорости.
