## Полоса замедления, как упругое тело

Полоса замедления ярко выраженя. В отличии от решения на сепаратрисе, она узкая и вытянутая в направлении от стенки. Можно отметить, что распределение скорости внутри полосы замеделния, ее форма и длина практически не меняются, она лишь изгибается то в лево то в право, как упругое тело. Возможно это только визуально так выглядит, но вероятно полоса замедления действительно состоит из одних и тех же частиц жидкости. Если бы с одной стороны в полосу замедления попадали бы новые частицы, то это тоже самое, что они бы резко замедлялись, в то время как с другой стороны от полосы частицы должны резко ускоряться, чтобы выйти из нее. Но я не вижу механизма, способно резко замедлять или ускорить частицы жидкости в продольном направлении. 

В качестве подтверждения, но не доказательства, что это так, хочу привести следующий рисунок. Не нем изображена серия сечений, параллельных стенке. Растояние от стенки указано под сечением. Цветом показана продольная компонента скорости. Линиями - транстверсальная (theta), сплошные линии - положительное значение (скорость направлена вправо), прерывистые - отрицательное (влево). Двигаться вперед по вермени - тоже самое, что двигаться по x в сторону убывания (движение сверху вниз в данном случае). 

<img src="xth_section_uw.png">

Если обратить внимание на полосу замедления и на то, какую транстверсальную скорость имеют частицы жидкости, из которых она состоит, то можно обнаружить подтверждение того, что полоса замедления и окружающее пространство не обменивается жидкостью. Там, где частицы жидкости смещаются вправо, транстверсальная скорость имеет полоижетльно значение, там где вправо - отрицательное. Можно отметить, что вне полосы замдления всегда существует второй максимум транстверсальной скорости. С чем он может быть связан?

Если считать полосу замедления упругим телом с условием непротикания на поверхности, то достаточно просто востановить поле скорости, вызваное ее движением.

## Аналогия с волнами на воде

Тут можно вспомнить простой пример с волнами на поверхности воды. Математически есть вихревой слой, с разных сторон от корого скорость частиц отличается. При волнах на воде отличается еще и плотность, но это уже не существенно в данном случае. Такое течнеие всегда не устойчиво, и на пальцах механизм его неустойчивости можно представить следующим образом. 

Представим, что жидкость покоится, а ветер имеет скорость. Жидкость и воздух разделяет вихревой слой. В равновестном состоянии вихревые линии распределны равномерно по поверхности, они компенсируют друг дурга и не вызывают никакого двжения жидкости в попересном направлении. Если теперь допустить неравномерность, то в той области, в которой сосредоточено больше вихревых линий, они вызывают движение поверхности воды. За этой областью жидкость поднимается, перед ней опускается. При этом поднявшаяся жидкость подгоняется ветром вперед, длина области повышеной завихрености сокращается, что приводит к фокусеровке завихрености и развитию неоднородности. 

Это известное объяснение, его я слышал на лекция по теории устойчивости, но может сейчас что-то переврал, так как восстанавливал из головы. 

Этот механизм объясняет в частности, почему возникающая волна имеет скорость среднюю между скоростью жидкости над поверхностью и под ней. А если слой смешения размыт и скорсть на нем меняется непрерывно, то скорсть воны совпадает со скорость точки, в которой она возникатеся. 

Можно педположить, что в нашем случае механизм неустойчивости тот же. В таком случае полосу ускорения должны окружать вихревые линии. Они действительно есть. На рисунке ниже изображена нормальная к стенке завихреность. В принципе ясно, как располагается полоса замедления - завихреность обрамляет ее.

## Нормальная скорость

Еще один интересный факт, что внутри полосы замедления всегда нормальная к стенке нормальная к стенке скорость v отрицательна (направлена от стенки), причем область отрицательной v достаточно точно совпадает со полосой замедления. 

На рисунке ниже на этот раз линиями показана нормальная к стенке скорость v. Прерывистые линии - отрицательные значение. Вне области отрицательных значений значения положительны, линии в этой области не нарисованы. 

<img src="xth_sect_uv.png">

Интересно, что минимальное значение v имеет там, полоса замедления сильнее всего отклоняется от положения равновесия, или может быть немножко раньше. Почему?

У основания полоса замедления практические не движется, как будто она закреплена. Это ествественно, так как в силу прилепания поперечные скорости у стенки малы, и перемещения которые они могли бы вызвать, тоже малы. Но это ведет к тому, что вихри, которые возникают в результате неустойчивости, нормальные к стенке, оказываются "зацепленными" за стенку и отклоняются в сторону, а не смещаются. При этом они дают вклад и в транстверсальную завихреность. Хотя наверное это не имеет значения, так как в решении на сепаратрисе этого нет. В нем полосы не так выражены, и в них нет выраженных нормальных к стенке поверхностей и соответствющих вихрей.
