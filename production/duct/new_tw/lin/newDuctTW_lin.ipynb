# Линейная устойчивость полосы замедления

В Новой бегущей волне, найденной в прямоугольном канале [duct_new_tw](duct/new_tw/duct_new_tw.ipynb), главная особенности - это полоса замедления, проходящая через центр расчетной области, смещающаяся то влево, то вправо (Рис. ниже). 

<img src="umap.png"> Рис.1. Бегущая волна при Re=1400. 

Как уже известно, волна замедления в данном случае испытывает синусоидальную неустойчивость. Волна замедления линейно неустойчива. Чтобы это продемонстрировать, можно рассмотреть недеформированную полосу замедления, направленную вертикально вверх, однородную вдоль трубы. Она, окажется линейно неустойчивой как раз к возмущениям такой формы. Кроме того, это возможность получить пульсационную составляющею движения в более простой форме. К сожалению, в отличии от решения на сепаратрисе, в данном случае полосу замедления нельзя выделить просто осреднением вдоль трубы, пульсации имеют слишком большую амплитуду (рис. ниже). Да и для решения на сепаратриcе осреднение дает полосу замедления в немного искаженном виде по той же причине. 

<img src="means.png" style="float:left"> Рис.2. Осредненное вдоль трубы поле скорости бегущей волны, Re=1400. Среднее поле сокрости совсем не похоже на полосу замедления с рис.1. 

Поле скорости, соответствующее недеформированной полосе замедления, придется создавать искусственно. В общем-то, должно быть не страшно не точно передать ее форму, так как интересующий нас механизм очень груб, и вероятно запустится на любой полосе замедления. Подтверждение этому то, что в решении на сепаратрисе полоса замедления имеет совсем другую форму, а механизм тот же (рис.ниже). 

<img src="edge_means.png" style="float:left"> Рис.3. Среднее поле скорости решения на сепаратрисе, Re=2200.

Можно даже не пытаться точно воспроизвести недеформированную полосу замедления, а вводить грубые схематизации. Так можно оценить, что важно для возникновения такого рода неустойчивости, а что не важно. Самая грубая схематизация, которая мне приходит в голосу, это поле скорости, в котором только две скорости, 0 и 1. 0 в полосе замедления - вертикальном прямоугольные в центре расчетной области, 1 везде вне ее. Но прежде чем вводить схематизации, стоит проверить полосу замедления, максимально приближенную к реальности. 

## Потеря устойчивости недеформированной полосой замедления

## Заключение

В общем, сконструированные искусственно полосы замедления оказываются подвержены синусоидальной неустойчивости, и возникающие в результате этой неустойчивости возмущения ведут к формированию продольных вихрей. Качественно картина воспроизводится верно, но количественно есть существенные различия. Вероятно это связанно с тем, что искусственная полоса замедления не совсем точно передать форму реальной. 

Полосу замедления приходится конструировать искусственно, так как ее нельзя в данном случае выделить просто осреднением по времени. 

Удобно ли в данном случае разделить поле скорости на стационарную и пульсационную составляющие? Пульсационную составляющею можно заменить решением линеаризованной системы, оно значительно проще. Получается два простых течения, во взаимодействии которых скрыт интересующий нас механизм. Их удобно исследовать. Но проблема в том, что среднее течение можно задать только искусственно. Кроме того, мне кажется такое разделение просто лишнее в данном случае.

Какое базовое течение мы бы не выбрали, оно не зависит от х. Основное слагаемое, дающее вклад в продольную завихренность ОХ выгляди так

$$
- u \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x}
$$

и посчитанное от базового течения, просто равно нулю. Нет разница, изучать ли вклад в формирование ОХ только от пульсаций ли от суммы пульсаций и базового течения, то есть самого исходного течения. Я думаю, можно не раскладывать исходное течение на среднюю и пульсационную составляющие движения, а рассматривать его таким, как оно есть. 

Замечание. Поперечные компоненты скорости во всех искусственных полосах замедления отсутствуют, равны 0.
