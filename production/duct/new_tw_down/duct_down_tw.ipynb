# Нижняя ветвь "новой" бегущей волны в канале

Была найдена "новая" бегущая волна, выделенная в плоском канале непосредственно из турбулентного течения. Лучшего названия я для него пока не придумал. Это решение не очень удобно для исследования так как амплитуда пульсаций в нем велика. 

В одной из статей было отмечено, что в каждом решении, которое они рассматривали, верхняя ветвь оказывается внутри области притяжения турбулентности в то время, как нижняя ветвь принадлежит ее границе, то есть сепаратрисе. Соответственно, только верхняя ветвь решения может находится внутри турбулентного аттрактора. 

Оказалось, что найденная бегущая волна действительно принадлежит верхней ветви решения. Продлевая его в сторону уменьшения числа Рейнольдса до точки бифуркации удалось получить решение на нижней ветви (Рис.1.). 

<img src="phase_space.png" style="float: left" width="50%"> Рис.1. Скорость сноса cf и число Рейнольдса  Re, при которых существует "новая" бегущая волны. Красная точка - Исходное решение, выделенное из турбулентного течения при Ре=1400. 

Решение на нижней ветви имеет значительно более низкую амплитуду пульсаций, чем на верхней. Это делает его идеальным кандидатом для исследования: решения является пристенным, система координат афинная, пульсации имеют небольшую величину. 

При продлении решения в пространстве параметров все параметры, кроме Re и cf, оставались постоянными

    Xmax * Ymax * Zmax = 5 x 1 x 1
    Im * Jm * Km = 64 x 40 x 32
    epsr = 0.25

Скорость сноса на верхней ветви стремится примерно к 0.73, на нижней к 0.8.

Можно проверить предположение о том, что продольная завихренность ох рождается в результате наклона нормальной к стенке завихренности oy. Для этого приведение рис.11. 

<img src="lin_oxoy.png"> Рис.11. Сравнение нормальной к стенке oy и продольной ox завихренности. Сечение то же, что и на предыдущем рисунке. Изолинии повторяют поле нормальной к стенке завихренности oy, приведенной на предыдущем рисунке. Цвет - продольная завихренность ox. Продольная завихренность ox действительно может возникнуть в результате наклона нормальной к стенке завихренности, так как области их существования совпадают, и ох достигает своего максимума там, где градиент скорости основного течения максимальный. Такая картина имеет место только в центре расчетной области. В области существования стационарных продольны вихрей механизм возникновения ox иной, как будет показано на следующем рисунке. 

<img src="lin_yconst_ox.png"> Рис.12. Параллельные твердой стенке сечения, как на рис.9. Цвет - продольная завихренность ox. В центре расчетной области ox сосуществует с oy, что может быть следствием наклона вихревых линий. Однако там, где образуются стационарные продольные вихри, продольная завихренность образуется, вероятно, за счет другого механизма, и возникает со сдвигом на пол периода.
