# Возникновение стационарных продольных вихрей

Принято считать, что продольнаые вихри, отвечающие за фозникновение полос повышенной и пониженной скорости, возрмируются в рзультате нелинейного взаимодействия пульсаций. Мы могли бы разобраться подробнее, в чем состоит это нелинейное взаимодействие. 

Исследуем бегущею волну, Re=2500, Xmax=5R. Для нее было показано ([TW-puls_to_ox-success](TW-puls_to_ox-success.ipynb) и [TW-puls_to_ox-continuation](TW-puls_to_ox-continuation.ipynb)),  что основной вклад в стационарную составляющею продольной завихрености $\Omega_x$ дает нелинейное взаимодействие пульсаций. Как формируется этот вклад?

## Общая идея

Можно разложить процесс на две составные части. Первое - среднее течение $\vec V$ оказывается линейно неустойчивым и на нем в рамках линейной теории фрмируются и нарастают возмущения определенной формы. Они подчиняются уравнению

$$
\frac{\partial \vec v}{\partial t} = - (\vec V, \nabla) \vec v - (\vec v, \nabla) \vec V - \nabla p + \frac{1}{Re} \Delta \vec v
$$


Второе - по решению, найденому в рамках линейной теории, посчитаем слагаемое $(\vec v, \nabla) \vec v$, отвечающее за нелинейное взаимодействие. Посмотрим, как бы менялось решение, если включить конвекцию. Это позволит определить эффект конвекции. 

Важно отметить, что решение, найденое в рамках линейной теории, мало отличается от пульсаций, наблюдаемых в действительности. Это говрит о том, что если включить нелинейность, то она не приведет к значительной перестройке решения. Достаточно просто посчитать нелинейное слагаемое по линейному решению, чтобы сделать прогноз о том, как бы выглядело решение полной системы.

Если считать, что завихреность движется от оси трубы к стенке, то в реальных пульсациях при этом положительная завихренность смещается в положительном направлении по углу, отрицательная - в отрицательном. В этом и заключается эффект, который приводи к возникновению продольной завихрености в среднем течении. Так как реальные пульсации получены путем вычитания среднего по времени течения, по мере того, как завихреность сносится в сторону, она пропадает для пульсационного течения, так как переходит в среднее течение. 

Такой эффект дает учет конвекции за счет угловой компоненты скорости. 

## Конвекция в угловом направлении

Смещение завихренности в угловом направлении можно получить, учтя конвективное слагаемое

$$
c^w_x = - w \frac{\partial \omega_x}{r \partial \theta}
$$



<img src="lin-w+om-map.png" style="float:left" width="680">  рис.6. Цветом показана угловая компонента скорости w, как на рисунке 4. Линии - линии уровня продольной завихрености ox. ox в основном сосредоточена между областями положительной и отрецательной w. Связяно это с тем, что именно w в первую очередь и порождает ox, а если уесть конвекцию, то тут же и сносит ox в сторону. 

<img src="cwx_map.png" style="float:left"> Рис.7. Распределение $c^w_x$. Всегда дает нужный эффет. Там, где завихрнеость отрицательна, она сносится в сторону увеличения $\theta$, при этом в области  $\theta < \pi/4$ значение завихрености увеличивается, в области $\theta > \pi / 4$ значение завихрнеости падает. Аналогично для положительной завихрнеости, только она сносится в сторону уменьшения $\theta$.

На следующем рисунке показана, какая часть продольной завихреность ox формируется за счет v, а какая за счет w. 

$$
ox\_v = - \frac{\partial v}{r \partial \theta}, \ \ \ \ \ \ \ \ ox\_w = \frac{\partial r w}{r \partial r}, \ \ \ \ \ \ \ \ ox = ox\_v + ox\_w
$$


<img src="ox_part_map.png">

Вклад в ox от v также хорошо синхронизирован с вкладом от w, и только усиливает завихрнеость в области существования полос. И это важно, так как если бы вся завихрнеость в области полос была образована за счет w, то совпадение положения продольных вихрей с местом образования полос было бы чисто случайностью, так как вытянуть жидкость от стенки они бы не могли, так как v для них = 0. 

Учет конвекции за счет w действительно позволяет полуить нужный эффект, но то дают другие слагаемые? 

## Эффект от остальных нелинейных слагаемых

Всего в уравнение для завихрнеости ox входит 6 нелинейных слагаемых, 3 конвекцтивных, 3 деформационных
$$
rot_x((\vec v, \nabla) \vec v) = - u\frac{\partial \omega_x}{\partial x} - v \frac{\partial \omega_x}{\partial r}  - w \frac{\partial \omega_x}{r \partial \theta} + \omega_x \frac{\partial u}{\partial x} + \omega_r \frac{\partial u}{\partial r} + \omega_\theta \frac{\partial u}{r \partial \theta}
$$

Я буду обращаться к ним, как к 

$$
c^u_x, c^v_x, c^w_x, d^1_x, d^2_x, d^3_x
$$ 


Так как $rot(-(\vec v, \nabla) \vec v) = rot(\vec v \times \vec \omega)$, сумма нелинейных слагаемых может быть вычислина существующим кодом из функции rp. На следющем графике приведено распределение этой величины.
