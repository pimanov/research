# Механизм образования стационарных продольных вихрей

Принято считать, что продольные вихри, отвечающие за формирование полос повышенной и пониженной скорости, формируются в результате нелинейного взаимодействия пульсаций. Мы могли бы разобраться подробнее, в чем состоит это нелинейное взаимодействие. 

Исследуем бегущею волну, Re=2500, Xmax=5R. Для нее было показано ([TW-puls_to_ox-success](TW-puls_to_ox-success.ipynb) и [TW-puls_to_ox-continuation](TW-puls_to_ox-continuation.ipynb)),  что основной вклад в стационарную составляющею продольной завихренности $\Omega_x$ дает нелинейное взаимодействие пульсаций. Как формируется этот вклад?

## Общая идея

Можно разложить процесс на две составные части. Первое - среднее течение $\vec V$ оказывается линейно неустойчивым и на нем в рамках линейной теории формируются и нарастают возмущения определенной формы. Они подчиняются линеаризованному уравнению

$$
\frac{\partial \vec v}{\partial t} = - (\vec V, \nabla) \vec v - (\vec v, \nabla) \vec V - \nabla p + \frac{1}{Re} \Delta \vec v
$$

Второе - по решению, найденному в рамках линейной теории, посчитаем слагаемое $(\vec v, \nabla) \vec v$, отвечающее за нелинейное взаимодействие. Посмотрим, как бы менялось решение, если включить конвекцию. Это позволит определить эффект конвекции, и выделить основные слагаемые, которые его дают. 

Важно отметить, что решение, найденное в рамках линейной теории, мало отличается от пульсаций, наблюдаемых в действительности. Это говорит о том, что если включить нелинейность, то она не приведет к значительной перестройке решения, и достаточно просто посчитать нелинейное слагаемое по линейному решению, чтобы сделать прогноз о том, как бы выглядело решение полной системы.

Если считать, что завихренность движется от оси трубы к стенке, то в реальных пульсациях при этом положительная завихренность смещается в отрицательном направлении по углу, а отрицательная - в положительном. В этом и заключается эффект, который приводи к возникновению продольной завихренности в среднем течении.

Такой эффект можно было бы получить, учтя конвекцию за счет угловой компоненты скорости w.

## Эффект нелинейных слагаемых

Всего в уравнения для завихренности ox входит 6 нелинейных слагаемых, 3 конвективных, 3 деформационных
$$
rot_x((\vec v, \nabla) \vec v) = - u\frac{\partial \omega_x}{\partial x} - v \frac{\partial \omega_x}{\partial r}  - w \frac{\partial \omega_x}{r \partial \theta} + \omega_x \frac{\partial u}{\partial x} + \omega_r \frac{\partial u}{\partial r} + \omega_\theta \frac{\partial u}{r \partial \theta}
$$

Я буду обращаться к ним, как к 

$$
c^u_x, c^v_x, c^w_x, d^1_x, d^2_x, d^3_x
$$ 


Так как $rot(-(\vec v, \nabla) \vec v) = rot(\vec v \times \vec \omega)$, сумма нелинейных слагаемых может быть вычислена существующим кодом из функции rp для проверки. На следующем графике приведены стационарные составляющие, как самых нелинейных слагаемых, так и их сумм. 

<img src="mean_nls.png">

Рис.8. Стационарная составляющая различных нелинейных слагаемых. Под каждым рисунком указано, что изображено и максимальное значение модуля величины.

1. Суммарный вклад нелинейных слагаемых. 
2. $c^v_x$ конвекция за счет радиальной компоненты скорости v.
3. $c^w_x$ конвекция за счет угловой скорости w.
4. Сумма $c^v_x + c^w_x$ - конвекция в поперечной плоскости.

6. $d^1_x$ растяжение вектора ox
7. Сумма $c^u_x+d^1_x$ - дает наиболее значительный вклад в образование вихрей. 
8. Сумма $d^2_x + d^3_x$ - еще два деформационных слагаемых, можно пренебрегать.

Под каждым рисунком приведено значение интеграла по сечению, отнесенного к интегралу суммы всех нелинейных слагаемых.

## Вклад конвекции вдоль трубы

Подавляющий вклад в формирование продольных вихрей дает сумма слагаемых $c^u_x+d^1_x$, отвечающих за конвекцию вдоль трубы растяжение продольной завихренности за счет деформации жидких частиц.

$$
c^u_x + d^1_x = -u\frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x}
$$

Ниже приведено распределение по сечениям трубы каждого из слагаемых и их суммы. Изображена только половина длины трубы. $c^u_x$ получается из $d^1_x$ сдвигом на четверть периода, и их сумма не зависит от продольной координаты. Это легко показать. 

<img src="sum_cuxd1x.png">

Фиксируем некоторые $(r,\theta)$, тогда $u$ и $\omega_x$ можно представить так 

$$
u = sin(x + \phi)
$$
$$
\omega_x = sin(x + \psi)
$$

Тогда
$$
-u \frac{\partial \omega_x}{\partial x} = - \sin(x + \phi) cos(x + \psi) 
$$

$$
\omega_x \frac{\partial u}{\partial x} = \sin(x + \psi) cos(x + \phi) = - cos(x + \psi + \pi/4) sin(x + \phi + \pi/4)
$$

То есть $c_u^x$ получается из $d^1_x$ просто сдвигом на $\pi/4$, а по формуле синуса суммы получаем

$$
- u \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x} = \sin(\psi - \phi)
$$

Сумма не зависит от x, а зависит только от разности фаз.

## Заключение

Конечно, и раньше было известно, что продольные вихри возникают в результате нелинейного взаимодействия пульсаций. Но этим методом можно показать не просто, какое слагаемое в уравнениях отвечают за это, а объяснить физику процесса, показать на пальцах, что происходит. Среди всех нелинейных слагаемых выявить наиболее существенные. 

Какую роль играют конвекция вдоль трубы? Она дает ярковыраженный положительный эффект при образовании продольных вихрей.
