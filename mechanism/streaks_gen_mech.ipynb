# О механизме поддержания вторичных течений

Пока я пытался понять физический смысл найденного нами механизма поддержания продольных вихрей, у меня возникло подозрение, что это не совсем тот механизм, который мы искали. Усиливая продольные вихри, он может снижать  интенсивность вторичного течения (поперечной компоненты среднего течения в исследуемых мной решениях). Пока я не понял, как это ясно изложить, но сейчас этого и не требуется. Здесь я хочу показать возможность сформулировать механизм поддержания вторичных течений без обращения к уравнениям эволюции завихренности. Механизм, который при этом вырисовывается, кажется простым и понятным. 

## Предварительные рассуждения

Для того, чтобы определить механизм поддержания вторичного течения, обратимся к уравнению Навье-Стокса непосредственно. Раньше я не знал, что делать с давлением в этом уравнении, и по этой причине выбрал для работы уравнение эволюции завихренности. Через давление движение жидкости в каждой точке оказывается связано со значением всех источниковых слагаемых во всей области течения. По этой причине, мне казалось, нельзя выделить какое-то одно слагаемое/механизм, формирующие ту или иную особенность движения. 

Сейчас я пришел к выводу, что эта проблема преодолима. Можно с каждым слагаемым связать свою часть давления, которая делает его бездивергентным. Иными словами, уравнение Навье-Стокса можно представить в виде:

$$
\frac{\partial \vec v}{\partial t} = \vec a_1 + \dots + \vec a_n - \nabla p = \vec b_1 + \dots + \vec b_n, 
$$

где 

$$
\mathrm{div} \vec b_i = 0
$$

и

$$
\vec b_i = \vec a_i - \nabla p_i
$$



Анализируя влияние не поле скорости $\vec v$ слагаемых $b_i$, а не $a_i$, мы не имеем дела с давлением, как с голобальным мехаищмом, который связывает значение всех источниковых слагаемых во всей области течения. Вместо этого, мы имеет дело с $p_i$, значение которого определяется только соответствующим $a_i$. Значение $p_i$ легко вычислить, но можно этого и не делать. Интегральный вклад $b_i$ в кинетическую энергию течения равен интегральному вкладу $a_i$, что позволяет установить роль $b_i$ в формироовании $\vec v$ без значния $p_i$:

$$
\int_D \vec b_i \cdot \vec v \ d\tau = \int_D \vec a_i \cdot \vec v \ d\tau
$$

(интегрирование выполняется по всей области течения). Кроме того, глядя на распределение $\vec a_i$ по различным сечениям трубы/канала тоже не сложно понять, как выглядит соответствующий $\vec b_i$. 

Таким образом, при анализе уравнений можно и не переходить к $b_i$, продолжая работать с $a_i$, но, мне кажется, приведенные тут рассуждения полезны для понимания правомерности таких дейсвтий.

## Теория

Рассмотрим решение в виде бегущей волны в плоском канале, как наиболее простое. Пусть $(x,y,z)$ -- продольная, нормальная к стенке и поперечная координаты, $\vec V = (U,V,W)$ и $\vec v' = (u',v',w')$ -- средняя и пульсационная составляющие движения. Осреднение выполняется по координате $x$. 

Уравнения баланса компонент вторичного течения имеют вид: 

$$
\frac{\partial V}{\partial t} = - U \frac{\partial V}{\partial x} - V \frac{\partial V}{\partial y} 
- W \frac{\partial V}{\partial z}  - \overline{u' \frac{\partial v'}{\partial x}}^x
- \overline{v' \frac{\partial v'}{\partial y}}^x - \overline{w' \frac{\partial v'}{\partial z}}^x
- \frac{\partial P}{\partial y} + \nu \nabla^2 V
$$


$$
\frac{\partial W}{\partial t} = - U \frac{\partial W}{\partial x} - V \frac{\partial W}{\partial y} 
- W \frac{\partial W}{\partial z}  - \overline{u' \frac{\partial w'}{\partial x}}^x
- \overline{v' \frac{\partial w'}{\partial y}}^x - \overline{w' \frac{\partial w'}{\partial z}}^x
- \frac{\partial P}{\partial z} + \nu \nabla^2 W
$$



Сложив первое уравнение, умноженное на $V$, со вторым, умноженным на $W$, получим уравнения баланса плотности кинетической энергии вторичного течения $e = (V^2 + W^2)/2$:

$$
\frac{\partial e}{\partial t} + (\vec V \cdot \nabla) e = 
- \underbrace{V \overline{u' \frac{\partial v'}{\partial x}}^x}_{A_1}
- \underbrace{V \overline{v' \frac{\partial v'}{\partial y}}^x}_{A_2}
- \underbrace{V \overline{w' \frac{\partial v'}{\partial z}}^x}_{A_3}
- \underbrace{W \overline{u' \frac{\partial w'}{\partial x}}^x}_{A_4}
- \underbrace{W \overline{v' \frac{\partial w'}{\partial y}}^x}_{A_5}
- \underbrace{W \overline{w' \frac{\partial w'}{\partial z}}^x}_{A_6} + \dots
$$

Отмечу, что вторичное течение бездивергентно, так как среднее течение не зависит от $x$:

$$
\frac{\partial V}{\partial y} + \frac{\partial W}{\partial z} = 0
$$



По этой причине конвективные слагаемые, градиент давления и вязкая диффузия не влияют на кинетическую энергию вторичного течения $E$,

$$
E = \int_D e d\tau
$$

Положительный вклад в поддержание $E$ могут дать только первые шесть слагаемых в правой части уравнения баланса $e$. 
Именно эти шесть слагаемых, обозначенные $A_i$, и будут нас интересовать.  

Для того, чтобы применить рассуждения из предидущего раздела, можно рассматривать в качестве $a_i$ такие величины:

$$
\vec a_1 = (0, - \overline{u' \frac{\partial v'}{\partial x}}^x, 0), \ \ \ \ 
\vec a_2 = (0, - \overline{v' \frac{\partial v'}{\partial y}}^x, 0), \ \ \ \ 
\vec a_3 = (0, - \overline{w' \frac{\partial v'}{\partial z}}^x, 0), \ \ \ \ 
$$

$$
\vec a_4 = (0,0,- \overline{u' \frac{\partial w'}{\partial x}}^x), \ \ \ \ 
\vec a_5 = (0,0,- \overline{v' \frac{\partial w'}{\partial y}}^x), \ \ \ \ 
\vec a_6 = (0,0,- \overline{w' \frac{\partial w'}{\partial z}}^x),
$$

а в качестве $\vec v$ поле скорости $(0,V,W)$. 

## Анализ решения в виде бегущей волны

Для бегущей волны, являющейся предельным решением на сепаратрисе, слагаемые $A_i$ имеют особенно простой вид. Анализ этого решения позволяет выявить основные детали механизма поддержания вторичного течения. 

Сразу уточню постановку задачи. Решение найдено в прямоугольной расчетной области с условием периодичности вдоль потока, условием прилипания на одной из боковых стенок и условием проскальзывания (зеркального отражения) на других боковых стенках. 

Для того, чтобы представить поле скорости этой бегущей волны, на рисунке ниже изображены продольная и поперечная компоненты среднего течения и амплитуда пульсаций в поперечном сечении канала $x = const$. Твердая стенка находится внизу. Основной особенностью этого решения в сравнении с другими решениями, которые я рассматривал, является то, что пульсации достигают максимальной амплитуды на наибольшем удалении от твердой стенки (в центре канала). 

<img src="pics/edge2000low_means_grey.png">

На следующем рисунке приведено значение слагаемых $A_i$ для этого решения. Среди всех слагаемых только два отличны от нуля -- это слагаемые $A_4$ и $A_6$. Другие слагаемые близки к нулю, так как в этом решении $v'$ близка к нулю во всей области течения (уступает $u'$ и $w'$ на 3 порядка). Можно показать, что при $v' = 0$ слагаемые $A_4$ и $A_6$ тождественно равны друг другу. Эти слагаемые дают положительный вклад в поддержание вторичного течения. 

<img src="pics/edge2000low_VWgen_grey.png"> 

Для того, чтобы объяснить согласованность фаз сомножителей в выделенных слагаемых, рассмотрим, какую форму имеют пульсации в этом решении. Три компоненты пульсационной составляющей движения в сечении $y=const$ приведены на рис. 3. Сечение совпадает с границей расчетной области, на которой пульсации имеют наибольшую амплитуду. Здесь и далее прерывистые линии соответствуют отрицательным значениям.

<img src="pics/edge2000low_puls_grey.png">

Схематически пульсационную составляющую движения можно представить в следующем виде: 

$$
u' = \beta \sin \alpha x \sin \beta z
$$
$$
v' = 0
$$
$$
w' = \alpha \cos \alpha x \cos \beta z
$$

при условии, что полоса пониженной скорости находится при $z=0$. Так заданное поле скорости удавлетворяет уравнению неразрывности. 

Посчитаем значение выделенных слагаемых по этому полю скорости

$$
A_4 = - W u' \frac{\partial w'}{\partial x} = W \alpha^2 \beta \sin^2 \alpha x \frac{\sin 2 \beta z}{2}
$$

$$
A_6 = - W w' \frac{\partial w'}{\partial z} = W \alpha^2 \beta \cos^2 \alpha x \frac{\sin 2 \beta z}{2}
$$

Их сумма имеет имеет простой вид:

$$
A_4 + A_6 = W \alpha^2 \beta \frac{\sin 2 \beta z}{2}
$$

Сомножители в каждом слагаемом оказываются в фазе, что обеспечивают наибольшую эффективность этому механизму. Согласованность фаз в $A_4$ является следствием уравнения неразрывности. Согласованность фаз в $A_6$ образуется сама собой. То есть это практически безусловно выполняется. Кроме того, сумма слагаемых не зависит от $x$, что позволяет опустить знак осреднения. По $z$ сумма слагаемых меняется с вдвое большей частотой, что соответствует тому, что пульсации имеют субгармонический тип. 

Теперь определим физический смысл выделенных слагаемых. На рисунке ниже изображена схема пульсационного движения. Мы находимся в сопутствующей системе отсчета, в которой поле скорости стационарно. Серым цветом в сечении $y=const$ обозначена полоса пониженной скорости. Черные окружности со стрелками -- линии тока пульсационной составляющей движения.

<img src="pics/puls_scheme.png">

Слагаемое $A_6 = - W \overline{w' \partial w'/\partial z}$ описывает следующий процесс. Масса жидкости, имеющая положительное значение $w'$, смещается в сторону больших $z$ (область А на рисунке ниже). Масса жидкости, имеющая отрицательное значение $w'$, смещается в сторону меньших значений $z$ (область Б на том же рисунке). Таким образом, образуется несимметрия в распределении пульсаций, которая переходит во вторичное течение. 

 <img src="pics/puls_scheme_zones.png"> 

Слагаемое $A_4 = - W \overline{u'\partial w' / \partial x}$ описывает растекание жидкости, вытекающей из областей А и области Б, вдоль оси х. 

Этот механизм выглядит простым и робастным. На рисунке ниже изображена схема этого процесса. Серым в поперечном сечении трубы обозначена полоса пониженной скорости. На полосе пониженной скорости возникают колебания, обозначенные красной стрелкой. Колебания происходят в плоскости, параллельной стенке ($v' = 0$). Эти колебания производят вторичные течения в той же плоскости, обозначенные зелеными стрелочками. Жидкость растекается из области, в которой происходят колебания, в сторону больших и меньших значений $z$. В силу наложенных условий симметрии и того факта, что колебания производят вблизи границы расчетной области, возникающие вторичные течения поворачивают в сторону стенки и вызывают движение жидкости от сетнки в центральную часть расчетной области, поддерживающее полосу пониженной скорости. 

<img src="pics/mech_scheme.png" >

Важной деталью такого объяснения является наличие границ расчетной области (условий симметрии), благодаря которым генерируемое вторичное движение принимает именно ту форму, которая обеспечивает подержание полос. Что может заставить жидкость двигаться аналогичным образом в реальном пристенном течении пока не ясно. 

## Анализ второго решения в виде бегущей волны

Рассмотрим другое решение в виде бегущей волны в геометрии течения в плоском канале. На рисунке ниже приведены средние характеристики течения. В этом случае пульсации достигают максимума внутри расчетной области. 

<img src="pics/edge2000up_means_grey.png">

На следующем рисунке приведено значение $A_i$ для этого решения. К выделенным в предыдущем решении слагаемым $A_4$ и $A_6$ в этом решении добавляются $A_1$ и $A_2$. Слагаемое $A_3$ дает отрицательный вклад в кинетическую энергию пульсаций, слагаемое $A_5$ вообще не имеет никаких особенностей, поэтому их я игнорирую.

<img src="pics/edge2000up_VWgen_grey.png">

Суммарный вклад $A_i$ в кинетическую энергию вторичного движения (интеграл $A_i$ по объему) приведен в таблице ниже.  

|$A_1$|$A_2$|$A_3$|$A_4$|$A_5$|$A_6$|$\Sigma A_i$|
|-----|-----|-----|-----|-----|-----|--------|
| 0.34| 0.06|-0.10| 0.25| 0.09| 0.36| 1.0    |

Для того, чтобы понять, почему распределения $A_i$ выглядят именно так, разберемся, как выглядят пульсации. 

Пульсации в центральной части расчетной области имеют такой же вид, как и в предыдущем решении, в частности $v'$ компонента в этой области близка к нулю. По этой причине все $A_i$, кроме $A_4$ и $A_6$, также близки к нулю. $A_4$ и $A_6$ в этой области тоже близки к нулю, так как в этой области близко к нулю $W$. Таким образом, в центральной части все $A_i$ близки к нулю. 

Пульсации вблизи границ расчетной области $y = const$, $z = const$ имеют меньшую амплитуду, чем в центре расчетной области, но именно они участвуют в поддрежании вторичного течения. На схеме ниже серый параллелепипед -- это расчетная область, черные линии со стрелками -- линии тока пульсационной составляющей движения на границах расчетной области. Твердая стенка находятся внизу. 

<img src="pics/3D_puls_scheme.png">

Линии тока в этом случае имеют практически такой же вид, как и в случае предыдущего решения, но проходят не только по границе $y=const$, но и по границам $z = const$. Соответственно, вблизи границ $z=const$ компонента $v'$ отличается от нуля. Механизм образования вторичных течений в этом случае тот же, что и в предыдущем, но если на границе $y=const$ за него отвечают слагаемые $A_4$ и $A_6$, то на границах $z=const$ -- слагаемые $A_1$ и $A_2$. Наличие границ (условий симметрии) в этом случае также играет существенную роль в образовании вторичных течений. 

## Заключение

Похоже, есть возможность иначе сформулировать механизм поддержания вторичных течения, без обращения к завихренности. Жидкость, вовлеченная в эти круговые движения, по инерции вытекает из области колебаний, образуя вторичные течения.
