## Теория

Уравнение для завихренности можно привести во многих формах, в частности нелинейное слагаемое может быть записано в одном из следующих видов

$$
d(v, om) =  - (\vec v, \nabla) \vec \omega + (\vec \omega , \nabla) \vec v = rot(\vec v \times \vec \omega) = - (\vec v, \nabla) \vec \omega + \Phi \vec \omega
$$

Но, если мы хотим предложить физическую интерпретацию процесса, нужно выбрать такой вариант записи уравнений, который ее допускает. В данном случае я выбираю форму записи 

$$
d(v, om) = - (\vec v, \nabla) \vec \omega + \Phi \vec \omega
$$

Здесь $\Phi -$ тензор скоростей деформаций. Завихренность переносится как примесь в жидкости за счет конвекции, и кроме этого поворачивается в пространстве за счет деформации жидкой частицы. Я хотел сказать, что конвекция не меняет направления вектора вихря, но это не совсем верно. Более правильно сказать, что конвекция не создает новой завихренности, а лишь переносит с потоком уже существующею. Тоже можно сказать и про деформационное слагаемое, оно тоже на создает новой завихренности, а лишь разворачивает в пространстве уже существующею. Разложим деформацию на сумму двух слагаемых $d = c + p$

Конвекция
$$
c(v,om) = - (\vec v, \nabla) \vec \omega
$$

Деформация
$$
p(v,om) = \Phi \vec \omega
$$

В проекции на ось трубы деформационное слагаемое p имеет следующий вид

$$
p_x(v,om) = \omega_x \frac{\partial u}{\partial x} + \omega_r \frac{1}{2} \left ( \frac{\partial u}{\partial r} + \frac{\partial v}{\partial x} \right) + \omega_\theta \frac{1}{2} \left( \frac{\partial u}{r \partial \theta} + \frac{\partial w}{\partial x} \right)
$$

Его снова можно разделить на два слагаемых $p_x = p_x^1 + p_x^2$

Вытягивание уже существующей завихренности за счет растяжения жидкой частицы
$$
p_x^1 = \omega_x \frac{\partial u}{\partial x}
$$

Пополнение продольной завихренности за счет поворота поперечной при деформации жидкой частицы
$$
p_x^2 = \omega_r \frac{1}{2} \left ( \frac{\partial u}{\partial r} + \frac{\partial v}{\partial x} \right) + \omega_\theta \frac{1}{2} \left( \frac{\partial u}{r \partial \theta} + \frac{\partial w}{\partial x} \right)
$$

Последнее слагаемое может быть вычислено и по другому

$$
p_x^2 = \omega_r \frac{\partial u}{\partial r} + \omega_\theta \frac{\partial u}{r \partial \theta} =
\frac{\partial v}{\partial x} \frac{\partial u}{r \partial \theta} - \frac{\partial w}{\partial x} \frac{\partial u}{\partial r}
$$


## Из чего складывается вклад от пульсаций?

Графики средних вдоль трубы

На первом графике суммарный вклад от пульсаций, такая картинка уже была, посчитан как

$$
\overline{d_x(\vec v',\vec \omega')}^x
$$

На втором - отдельно конвективное слагаемое

$$
\overline{c(\vec v',\vec \omega')}^x
$$

На третьем - вытягивание завихренность за счет растяжения частицы

$$
\overline{p1(\vec v', \vec \omega')}^x
$$

На четвертом
$$
\overline{p2(\vec v', \vec \omega')}^x
$$

Над графиками приведено значение amp. Синему цвету соответствует значение -amp, красному +amp. Все картинки в едином масштабе по цвету.

Ниже графика приведено максимальное значение, которое достигает изображенная функция. 

Получается так, что главную роль в образовании завихренности играет конвекция. Так как пульсации практически отсутствуют на биссектрисе, вероятно завихренность переносится от оси трубы к стенке в радиальном направлении. 

слагаемое p1 также дает значительный положительный вклад, в 2 раза меньший, чем конвекция. Но это не очень существенно. Это слагаемое не создает движения в поперечной плоскости. Оно отвечает за то, что в тех точках, где $\partial u / \partial x$ больше нуля, в силу уравнения неразрывности, в поперечной плоскости частицы жидкости приближаются к этой точке. При это их поперечная скорость не меняется, но завихренность увеличивается, так как они становятся ближе друг к другу. 

Слагаемое p2, соответствующее вкладу поперечной завихренности, пренебрежимо мало. Я вычислял все эти слагаемые очень грубо, в лоб в соответствии с формулами, не обращая внимание на перемежаемость сетки. Величина слагаемого p2 оказывается сравнима с той ошибкой, которую я в итоге получил. Оно оказывается сравнимо с величиной (c + p1 + p2 - d), которая должна быть равна 0. 

## Составляющие конвекции

Первый график ниже - вклад в ох от конвекции, посчитанный как

$$
\overline{c_x(\vec v',\vec \omega')}^x
$$

Это слагаемое можно разложить на компоненты, которые изображены на следующих графиках
$$
c_x(\vec v, \vec \omega) = - u \frac{\partial \omega_x}{\partial x} - v \frac{\partial \omega_x}{\partial r} - w \frac{\partial \omega_x}{r \partial \theta}
$$

Вторая кртинка
$$
cx = \overline{ - u' \frac{\partial \omega_x'}{\partial x}}^x
$$

Третья картинка
$$
cr = \overline{ - v' \frac{\partial \omega_x'}{\partial r}}^x
$$

Четвертая картинка
$$
ct = \overline{ - w' \frac{\partial \omega_x'}{r \partial \theta}}^x
$$

## Альтернативное разложение конвекции на слагаемые

Конвективное слагаемое можно разложить на слагаемые и по другому, пользуясь соотношением 

$$
c_x(\vec v, \vec \omega) = - \frac{\partial u \omega_x}{\partial x} - \frac{\partial r v \omega_x}{r \partial r} - \frac{\partial w \omega_x}{r \partial \theta}
$$

Следующие графики построены в соответствии с этим разложением

На первом графике снова суммарный вклад от конвекции

$$
\overline{c_x(v',om')}^x
$$

Вторая картинка
$$
cx = \overline{ - \frac{\partial  u' \omega_x'}{\partial x}}^x = 0
$$

Третья картинка
$$
cr = \overline{ - \frac{\partial r v' \omega_x'}{r \partial r}}^x
$$

Четвертая картинка
$$
ct = \overline{ - \frac{\partial w' \omega_x'}{r \partial \theta}}^x
$$

<img src="tw-conv_part_1.png" align="left">

Такое представление удобнее, так как cx дает в этом случае просто 0. Интеграл от cr по радиусу тоже дает 0, что позволяет смотреть на конвекцию именно как на перенос завихренности без создания новой. cr и ct дают только положительный вклад в основной вихрь.
