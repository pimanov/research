## Вклад в энергию продольной завихренности OX

На следующем графике приведен вклад в энергию средней завихренности $\Omega_x^2$ (Я к сожалению забыл то слово, которым она называется). Стационарную составляющую продольной завихренности $\Omega_x$ здесь, как и раньше, я часто буду обозначать просто OX. 

В первой строке суммарный вклад от стационарной и пульсационной составляющих течения. Вторая строка - вклад только от стационарной составляющей. Третья - только от пульсационной. 

Первая колонка - суммарный вклад от всех нелинейных слагаемых, приведенных в следующих колонках.

Вторая колонка - вклад слагаемого (cux+d1x), умноженных на OX
$$
cux + d1x = - u \frac{\partial \omega_x}{\partial x} + \omega_x \frac{\partial u}{\partial x}
$$

Третья колонка - вклад конвекции в поперечной плоскости (cvx + cwx), умноженной на OX
$$
cvx + cwx = - v \frac{\partial \omega_x}{\partial r} - w \frac{\partial \omega_x}{r \partial \theta}
$$

Четвертая колонка - деформационные слагаемые OX*(d1x + d2x)
$$
d1x + d2x = \omega_r \frac{\partial u}{\partial r} + \omega_\theta \frac{\partial u}{r \partial \theta}
$$

Так как в данном случае нас интересует стационарная составляющая завихренности, все слагаемые осреднены вдоль трубы.

Под каждой картинкой указано, какой относительный вклад (интеграл по сечению трубы) оно дает в энергию продольной завихренности

Как и раньше, основной вклад дает слагаемое cux+d1x пульсационной составляющей движения. 
Среднее течение в этом случае вовсе можно игнорировать. Конвекция вдоль трубы и деформации d2x+d3x для него равны нулю точно.
