# Как формируется продольная завихренность

Принято считать, что продольная компонента завихренности возникает в результате нелинейного взаимодействия турбулентных пульсаций. Это слишком обще, и конечно есть какое-то механическое объяснение этого процесса, объяснение на пальцах. Исследуя наше решения мы могли бы выявить этот механизм, и это было бы большое достижение!

### Уравнения для вектора вихря $\vec \omega$


Обратимся к уравнению для завихренности

$$
\frac{\partial \vec \omega}{\partial t} = - (\vec v , \nabla) \vec \omega + (\vec \omega , \nabla) \vec v + \nu \Delta \vec \omega
$$

Здесь производная завихренности по времени складывается из трех частей: конвекция, деформация вихревых линий и вязкая диссипация. Слагаемое для деформации можно переписать и в другом виде, через тензор скоростей деформаций $\Phi$, что передает его физический смысл

$$ 
(\vec \omega , \nabla ) \vec v = \Phi \vec \omega
$$

### Уравнения на продольную компоненту завихренности $\omega_x$


Проектируя уравнение на ось трубы получаем уравнение на продольную компоненты завихренности $\omega_x$

$$
\frac{\partial \omega_x}{\partial t} = - u \frac{\partial \omega_x}{\partial x} - v \frac{\partial \omega_x}{\partial r} - \frac{w}{r} \frac{\partial \omega_x}{\partial \theta} + \omega_x \frac{\partial u}{\partial x} + \omega_r \frac{\partial u}{\partial r} + \frac{\omega_\theta}{r} \frac{\partial u}{\partial \theta} + \nu \Delta \omega_x
$$

Если мы интересуемся процессом возникновения $\omega_x$, то интерес представляют только два слагаемых, это

$$ 
\omega_r \frac{\partial u}{\partial r} + \frac{\omega_\theta}{r} \frac{\partial u}{\partial \theta} 
$$

Если продольная завихренность равна нулю, то все другие слагаемые тоже равны нулю, и за ее возникновение могут отвечать только два приведенных. По своей сути все другие слагаемые отвечают за перераспределение уже существующей завершенности. 

Если поле скорости не зависит от продольной координаты х, то эти два слагаемых в сумме в точности равны нулю. Соответственно, если в потоке не было продольной завихренности и поле не меняется вдоль трубы, то она и не возникнет вообще. Следовательно неоднородность вдоль трубы имеет первостепенное значение в нашем случае. 

$$
\omega_r \frac{\partial u}{\partial r} = \left( \frac{\partial u}{r \partial \theta} - \frac{\partial w}{\partial x} \right)  \frac{\partial u}{\partial r} = \frac{\partial u}{r \partial \theta}  \frac{\partial u}{\partial r} - \frac{\partial w}{\partial x}   \frac{\partial u}{\partial r}
$$

$$
\omega_\theta \frac{\partial u}{r \partial \theta} = \left( \frac{\partial v}{\partial x} - \frac{\partial u}{\partial r} \right)  \frac{\partial u}{r \partial \theta} =  - \frac{\partial u}{r \partial \theta} \frac{\partial u}{\partial r}  + \frac{\partial v}{\partial x} \frac{\partial u}{r \partial \theta}
$$

Чтобы оценить вклад этих слагаемых, я отбрасываю общею часть и буду вычислять только вторую их часть, которая зависит от х, а это 

$$
d_2 = -\frac{\partial w}{\partial x} \frac{\partial u}{\partial r}
$$
$$
d_3 = \frac{\partial v}{\partial x} \frac{\partial u}{r \partial \theta} 
$$

Иначе они теряются на фоне однородной вдоль трубы части. Я буду обращаться к ним, как к $d_2$ и $d_3$.

Эти же два слагаемых можно записать через тензор скоростей деформаций

$$
\omega_r \frac{1}{2} \theta_3 = 
\left( \frac{\partial u}{r \partial \theta} - \frac{\partial w}{\partial x} \right) \frac{1}{2} \left ( \frac{\partial u}{\partial r} + \frac{\partial v}{\partial x} \right)
$$

$$
\omega_\theta \frac{1}{2} \theta_2 = \left( \frac{\partial v}{\partial x} - \frac{\partial u}{\partial r} \right) \frac{1}{2} \left( \frac{\partial u}{r \partial \theta} + \frac{\partial w}{\partial x} \right)
$$

В них также присутствует общая часть, совпадающая по величине но противоположная по знаку, однородная вдоль трубы. Если ее отбросить, то окажется, что оба слагаемых вычисляются по одной и той же формуле

$$
d_2^{\Phi} = d_3^{\Phi} = \frac{1}{2} \left( \frac{\partial v}{\partial x} \frac{\partial u}{r \partial \theta} - \frac{\partial w}{\partial x} \frac{\partial u}{\partial r} \right)
$$

Я буду обращаться к ним, как к $d_i^{\Phi}$. Верно, что

$$
d_2^{\Phi} = d_3^{\Phi} = \frac{d_2 + d_3}{2}
$$

так что я буду строить графики для $d_2$, $d_3$, $d_2+d_3$. 

### Физический смысл

В идеальной жидкости удобно рассматривать вихревые линии, так как они оказываются вморожены в жидкую среду, а вихревые трубки сохраняют свою интенсивность. В нашем случае есть вязкость, и глобально вихревые линии больше не сохраняются, но слагаемое $(\vec \omega, \nabla) \vec v$ в уравнении для завихренности по-прежнему отвечает за поворот и деформацию уже существующих в потоке вихревых линий. 

Соответственно, $d_2^{\Phi}$ и $d_3^{\Phi}$ отвечают за то, как компоненты вектора завихренности $\omega_r$, $\omega_\theta$ поворачиваются и приобретают $x$ составляющею за счет деформации жидкой частицы. Так как эти два слагаемых равны, нет возможности выделить только одну компоненту поперечной завихренности, которая бы давала начало продольной ее компоненте. 

Среди слагаемых $d_2$, $d_3$ можно найти, какое слагаемое где дает больший вклад, но как это потом интерпретировать - не очень понятно. 

Можно умножить все на $\omega_x$ и получить уравнения на энергию, но только не всего вектора вихря, а только продольной его компоненты.

### Представление в плоскости поперечного сечения


В плоскости поперечного сечения локально в каждой точке можно перейти к новым ортонормированным координатам (p,q), в них выражение для d преобразуется к виду

$$
d = d_2 + d_3 = \frac{\partial v_p}{\partial x} \frac{\partial u}{\partial q} - \frac{\partial v_q}{\partial x} \frac{\partial u}{\partial p}
$$

Cистема координат $(p,q,x)$ получина поворотом системы координат $(r,\theta,x)$ (в формуле я только поеменял $r$ на $p$, $\theta$ на  $q$); $v_p, v_q - $ проекции вектора скорости на соответствующие направления. 

Если теперь выбрать такую систему координат, в которой $q$ соответствует направлению наискорейшего роста $u$, то останется слагаемое только с $\partial u / \partial q$, при этом

$$ 
\frac{\partial u}{\partial p} = 0
$$

$$
d = \frac{\partial v_p}{\partial x} \frac{\partial u}{\partial q}
$$

Это полезно иметь ввиду, глядя на картинки поперечного сечения трубы. Вектор $v_p$ оказывается касательным к линиям равного уровня $u$. Направление же градиента скорости $u$ перпендикулярно этим линиям. Получилась простая и наглядная формула, но она не отражает физики процесса. 




С точки зрения физики можно предложить следующею интерпретацию. Спроектируем вектор вихря в поперечную плоскость сечения. Пусть s - направление этой проекции в плоскости сечения. Обозначим ее $\omega_s$. За счет деформации жидкой частицы она поворачивается в пространстве так, что дает вклад в продольную завихренность

$$
d = \omega_s \theta^*
$$

Здесь $\theta^*$  - соответствующая компонента тензора скоростей деформаций. На нее можно также смотреть, как на компоненту двумерного тензора - проекции тензора скоростей деформаций в плоскость, построенную на направлениях $(s,x)$. Таким образом весь процесс является двумерным и происходит в плоскости $(s,x)$.

$$
\theta^* = \frac{1}{2} \left( \frac{\partial u}{\partial s} + \frac{\partial v_s}{\partial x} \right)
$$

Направление s в общем случае не совпадает с направлением q. 

Таким представлением я еще не пользовался, но оно может быть очень продуктивным для понимания процесса. Можно построить серию поперечных сечений, в которых изобразить вектора $\omega_s$, первое и второе слагаемое от $\theta^*$.
