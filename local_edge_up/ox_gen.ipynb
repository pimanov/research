# Механизм генерации продольных вихрей на верхней ветви

При исследовании исходного решения, принадлежащего сепаратрисе, был выделен некоторый механизм, ответственный за образование продольных вихрей. Интересно выяснить, какую роль выделенный механизм играет в решении с верхней ветви. Для этого вновь обратимся к уравнению для стационарной составляющей продольной завихренности $\Omega_x$:

$$
\frac{\partial \Omega_x}{\partial t} - \nu \nabla^2 \Omega_x = c_f \frac{\partial \Omega_x}{\partial x} - (V, \nabla) \Omega_x + (\Omega, \nabla) V_x - \overline{(v', \nabla) \omega'_x} + \overline{(\omega', \nabla) v'_x}
$$

Поле стационарной продольной завихренности $\Omega_x$ (обозначаемое также OX) изображено на рисунке ниже. Как и на нижней ветви, распределение OX соответствует паре продольных вихрей, хотя и имеет менее правильную форму. 

<img src="OXgen_img/OX.png">

На графике ниже черной прерывистой линией изображено распределение вдоль трубы величины $\Omega_x^2$, осредненной по сечению трубы. Также синей линией приведен средний по сечению трубы вклад в $\Omega_x^2$ выделенных слагаемых

$$
-\overline{u' \frac{\partial \omega_x'}{\partial x}} + \overline{\omega_x' \frac{\partial u'}{\partial x}},
$$

а красной -- вклад других слагаемых из правой части уравнения. Вклад выделенных слагаемых оказывается положительным, в то время как вклад других слагаемых -- отрицательным (в среднем), что позволяет утверждать, что и в этом случае выделенные слагаемые играют ключевую роль в формировании продольных вихрей. Амплитуда $\Omega_x^2$ уменьшена в 20 раз, чтобы приведенные величины были сравнимы.

Стоит подумать о том, чтобы разделить правою часть уравнения на слагаемые аккуратнее, чтобы за каждым из слагаемых остался собственный физический смысл. 

Ниже немного подробнее рассмотрен вопрос о том, какие из оставшихся слагаемых компенсируют выделенные. 

## Компенсация выделенных слагаемых

Можно предположить, что слагаемые $\overline{-u' \partial \omega'_x / \partial x}$ компенсируются оставшимися слагаемыми в выражении $-\overline{(v', \nabla) \omega_x'}$, и, аналогично, слагаемое $\overline{\omega' \partial u' / \partial x}$ компенсируется оставшимися слагаемыми в выражении $\overline{(\omega', \nabla) u'}$. На графиках ниже изображены компоненты слагаемых $-\overline{(v', \nabla) \omega_x'}$ (слева) и $\overline{(\omega', \nabla) u'}$ (справа), домноженные на $\Omega_x$. Синяя кривая соответствует выделенным слагаемым, красная -- паре оставшихся слагаемых (по одному я их не считал). Видно, что они неплохи компенсируют друг друга, но иначе вряд ли могло бы быть, так как все слагаемые, не попавшие на графики ниже, считаются по стационарной составляющей движения, и с одной стороны они меньше по амплитуде, с другой -- представляют другой род движения. 

<img src="OXgen_img/noted_mean_cmp.png">

Не только средние по сечению трубы величины демонстрируют компенсацию, но и в сечении это тоже верно. Ниже приведены компоненты $-\overline{(v', \nabla) \omega_x'}$, соответствующие производной вдоль трубы (верхний ряд) и производной в поперечной плоскости (нижний). Они в некоторой степени повторяют друг друга по форме. Однако, это может быть лишь следствие того, что сами пульсации распределены таким же образом и их распределение задает распределение всех величин, посчитанных по ним.

В общем, приведенные графики не позволяют что-то утверждать о том, какие именно слагаемые компенсируют выделенные, но важно то, что если попробовать откинуть ту часть выделенных слагаемых, которая компенсируется, то влияние того, что останется, может оказаться вовсе не таким большим, и оно может быть сравнимо с другими слагаемыми. Что именно отбрасывать не известно, но можно сложить все слагаемые, посчитанные по пульсационной составляющей движения, и сравнить их с тем, что дают слагаемые, посчитанные по среднему течения. На графике ниже первым соответствует синяя линия, а вторым -- красная. Вторые оказываются не просто сравнимы с первыми, но и дают практически такой же средний вклад в образование $\Omega_x^2$. Однако, при х ~ 3, в центре области генерации $\Omega_x$, пульсации наиболее эффективны в образовании $\Omega_x$, в то время, как среднее течение оказывает негативное влияние, и оно связано не только и не сколько с конвекцией вдоль трубы, а в первую очередь со слагаемым $\Omega_x \partial U/\partial x$, то есть с поворотом возникающих продольных вихрей.

## Согласованность пульсаций продольной скорости и завихренности

Каким образом формируется $\omega_x'$? По-видимому, таким же, как и раньше: наибольший вклад дает слагаемое, связанное с поворотом нормальных к стенке вихрей в присутствии радиального градиента скорости, но ту часть $\omega'_x$, которая участвует в поддержании $\Omega_x$, создает слагаемое $\Omega_x \partial u'/\partial x$. Оставшиеся слагаемые в правой части уравнения 

$$
\frac{\partial \omega'_x}{\partial t} - \nu \nabla^2 \omega_x' = c_f \frac{\partial \omega_x'}{\partial x} - (V,\nabla)\omega_x' - (v', \nabla) \Omega_x - ((v', \nabla) \omega_x')' + (\Omega, \nabla) u' + (\omega', \nabla) U + ((\omega', \nabla) u')'
$$

также дают существенный вклад в \omega_x', и не ясно, важен ли он при формировании $\Omega_x$. 

Мне до конца не понятно, как ясно показать, какие именно слагаемые создают ту часть $\omega_x'$, которая участвует в формировании $\Omega_x$. Представление о том, что именно слагаемые $\Omega_x \partial u' / \partial x$ является основным, возникло у меня при изучении бегущей волны в плоском канале. Там различные слагаемые создают $\omega_x'$ в разных областях пространства, и на месте продольных вихрей отличалось от нуля только это слагаемое. В случае решение, исследуемого здесь, это не так. Тем не менее, здесь будут приведены аналогичные картинки. 

Будем считать, что среди двух слагаемых, ответственных за генерацию $\Omega_x$, ведущим является слагаемое $\overline{\omega_x' {\partial u'}/{\partial x}}$, так как механизм генерации $\omega_x'$ обеспечивает согласованность фаз именно его сомножителей.  Ниже на рисунке изображено слагаемое $\overline{\omega_x' {\partial u'}/{\partial x}}$ в первой строке. В строках ниже изображен вклад слагаемых, ответственных за генерацию $\omega_x'^2$. Во второй строке приведен вклад слагаемого $\Omega_x \partial u'/\partial x$. Его форма повторяет форму слагаемого в первой строке, что в некоторой степени подтверждает, что именно оно создает ту часть $\omega_x'$, которая участвует в создании $\Omega_x$. В третьей строке приведен вклад в $\omega_x'^2$, связанный с поворотом нормальных к стенке вихрей $\omega_r'$ в присутствии $\partial U / \partial r$. К генерации $\Omega_x$ он, по-видимому, отношения не имеет. В третьей строке приведен вклад всех остальных слагаемых. Они также имеют существенную амплитуду, и мне сейчас не ясно, какое отношение они имеют к генерации $\Omega_x$.
