## Линейная устойчивость среднего течения на верхней ветви решения

Среднее течение $(U,V,W)$ оказывается линейно неустойчивым. Ниже на графике представлена амплитуда возмущения, развивающегося на нем в рамках линеаризованных уравнений на него (синяя сплошная линия). После некоторого периода установления возмущение выходит на экспоненциальный рост. Инкримент нарастания $\lambda = 0.05$. Пунктирная линия представляет функцию $a = exp(-7.5 + 0.05t)$. Наиболее быстрорастущее собственно возмущение, на которое выходит решение линеаризованных уравнений при $t > 200$, я буду обозначать $\mathbf v_{lin}$.

<img src="lin_amp.png">

Поле скорости $\mathbf{v}_{lin}$ является периодическим по времени с периодом $T = 34$ (Если разделить его на $exp(-7.5 + 0.05t)$). Осредняя по времени квадрат скорости решения $\mathbf{v}_{lin}$, можно получить амплитуду этого движения. На графике ниже изображены средние по сечению трубы амплитуды $\mathbf{v}_{lin}$ (красная линия) и $\mathbf v'$ -- наблюдаемой в действительно пульсационной составляющей движения (синяя линия). Их распределения плохо совпадают. Это объясняется тем, что $\mathbf{v}_{lin}$ не имеет ничего общего и $\mathbf{v'}$.

## Второе собственное решение линеаризованной системы уравнений

Хотя $\mathbf{v}_{lin}$ не соответствует наблюдаемым пульсациям, по видимому, есть и другие собственные возмущения среднего течения с положительным инкрементом нарастания. Такой вывод можно сделать, глядя на первый график, на котором изображено, как растет амплитуда линеаризованного возмущения. В период времени $50 < t < 150$ решение, по видимому, близко к другому собственному возмущению, которое также нарастает, хотя и медленнее, чем $\mathbf{v}_{lin}$. Я буду назвать это решение $\mathbf{w}_{lin}$. Инкремент его нарастания $\lambda = 0.017$. Оно также является периодическим по времени с периодом $T = 30$, что близко к периоду возмущений, наблюдаемых в исследуемом решении. 

На графике ниже приведена средняя по сечению амплитуда $\mathbf w_{lin}$ в сравнении со средней амплитудой $\mathbf v'$. Их распределения хорошо совпадают. Амплитуда была посчитана на периоде изменения этого решения от t=100 до t=130. На этом интервале решение достаточно близко к периодическому, хотя можно видеть, как на его фоне развивается решение $\mathbf v_{lin}$

<img src="lin_img/lin0_amp.png">

На рисунке ниже изображена амплитуда $\mathbf{w}_{lin}$ в тех же сечениях, что и раньше. Иллюстрация подтверждает, что $\mathbf{w}_{lin}$ воспроизводит $\mathbf v'$. 

<img src="lin_img/lin0_mean_amp.png">

Нет ничего удивительного в том, что прежде чем выйти на режим, соответствующий решению $\mathbf{v}_{lin}$, я попадаю на режим $\mathbf{w}_{lin}$. Это связанно с тем, что в качестве начального поля скорости я выбирал $\mathbf{v'}$. Может так быть, что я обманываюсь, и при 50 < t < 150 режим просто не успевает перестроиться, поэтому решение $\mathbf{w}_{lin}$ похоже на $\mathbf{v'}$. Но это решение показывает достаточно уверенный экспоненциальный рост и меняется достаточно близким к периодическому образом. Я думаю, есть алгоритмы, которые позволяют посчитать второе или третье по величине инкремента собственные решения, не находя их все. И эти алгоритмы с вычислительной точки зрения не сильно сложнее того, каким образом я ищу первое собственное решение. Может быть стоит найти вторые и третьи собственные решения, чтобы явно показать, что среди них есть (или нет) то, которое соответствует наблюдаемым пульсациям.

## Устойчивость поля скорости (U,0,0)

Если теперь занулить поперечное движение в стационарном течении, то такое поле скорости $(U,0,0)$ оказывается устойчиво. 

<img src="lin_img/amp_time_VWz.png">

Возникающие возмущения являются периодическим по времени с периодом $T = 201$. Очевидно, что с таким периодом по времени оно не может быть похоже на $\mathbf v'$.
