# Два механизма возникновения производной вдоль потока. 

Мы нашли, что в тех нескольких случаях, которые мы рассматривали, за образование стационарной составляющей продольной завихренности отвечают слагаемые:

$$
\frac{\partial \Omega_x}{\partial t} = - \overline{u' \frac{\partial \omega'_x}{\partial x}} + \overline{\omega'_x \frac{\partial u'}{\partial x}} + ... 
$$

Физический механизм, который они представляют, по-видимому, состоит в том, что жидкие частицы растягиваются и сжимаются вдоль трубы. Растяжение частиц вдоль трубы ведет к образованию производных по х. Но производные по х может дать и другой механизм, связанный с тем, что линии тока в течении не прямые, а как-то изгибаются. Там, где они проходят под углом к оси трубы, при вычислении производной по х, как разности значений в двух близких точках, значение в одной точке может быть с одной линии тока, а значение в другой -- с другой. При этом на разных линиях тока значения могут быть различными, при том, что вдоль самих линий тока они не меняются. Следовательно производная по х может возникнуть без растяжения частиц. 

Приведенные рассуждения не идеальны, но как их сделать лучше, мне пока не ясно. Главное, что я хотел сказать, что производная вдоль трубы может возникать в следствии двух механизмов, и только первый из них, связанный с изменением длины жидких частиц, может вести к образованию продольной завихренности. Однако вклад в $\Omega_x$ через выделенные слагаемые дают оба механизма. Мне кажется, что со стороны других слагаемых можно выделить те, которые вклад второго механизма в точности компенсируют. 

На мысль о том, что они существуют, наводит тот факт, что вклад других источниковых слагаемых, как правило, имеет отрицательное значения на месте продольных вихрей. Кроме того, его форма повторяет некоторые особенности вклада выделенных слагаемых с противоположным знаком, которые пропадают после сложения.  

То, что я к настоящему моменту придумал, это разделить производную по х на две составляющие: производную вдоль вектора скорости и производную поперек него. Предполагается, что производная вдоль вектора скорости связанна с изменением длины частицы, которая по ней движется. Производная поперек вектора скорости с изменением длины частицы не связана. Конечно, это не совсем так, и разделение следовало бы провести более аккуратно, но такой подход уже позволяет разделить первое из выделенных слагаемых на две части, вторая из которых в точности компенсируется соответствующими частями других нелинейных слагаемых. 

## Формула для компонент производной по х. Рассуждения на плоскости. 

Выберем произвольную точку пространства. В ней жидкость движется с некоторой скоростью $\mathbf v$. Рассмотрим плоскость, в которой лежит ось х и вектор v (см. рисунок). Введем в этой плоскости вторую ост координат s перпендикулярную х, так что вектор скорости раскладывается на компоненты $(v_x, v_s)$. Также введем вторую систему координат (g,f), где ось g сонаправлена с направлением скорости, f перпендикулярна g. 

<img src="math/2Dscheme.png">

Производная вдоль трубы складывается из производной вдоль направления скорости и производной поперек скорости в соответствии с формуле:

$$
\frac{\partial}{\partial x} = \frac{\partial g}{\partial x} \frac{\partial }{\partial g} + \frac{\partial f}{\partial x}\frac{\partial}{\partial f}
$$

Запомним, что первое слагаемое соответствует производной вдоль потока, а второе -- поперек, и вернемся в исходную систему координат (x,s):

$$
\frac{\partial}{\partial x} = \frac{\partial g}{\partial x} \left( \frac{\partial x}{\partial g} \frac{\partial }{\partial x} + \frac{\partial s}{\partial g} \frac{\partial }{\partial s} \right) + \frac{\partial f}{\partial x} \left( \frac{\partial x}{\partial f}\frac{\partial}{\partial x} + \frac{\partial s}{\partial f}\frac{\partial }{\partial s}\right)
$$

Так как система координат (g,f) определяется направлением вектора скорости, коэффициенты перехода к ней и обратно выражаются через компоненты этого вектора. Выражение доля производной по х приобретает вид:

$$
\frac{\partial}{\partial x} = \left( \frac{v_x^2}{\mathbf v^2} \frac{\partial }{\partial x} + \frac{v_x v_s}{\mathbf v^2} \frac{\partial }{\partial s} \right) + \left( \frac{v_s^2}{\mathbf v^2}\frac{\partial}{\partial x} - \frac{v_x v_s}{\mathbf v^2}\frac{\partial }{\partial s}\right)
$$

Раскрыв скобки, получаем тождество, но нам важно, что первая скобка соответствует производной по потоку, а вторая -- поперек. 

## Выведение формулы для компонент производной по х. Рассуждения в пространстве. 

Вернемся в трехмерное пространство. Пусть (x,y,z) -- Декартова система координат. Ось s лежит в плоскости, образуемой осями (y,z).

<img src="math/3Dscheme.png">

Учитывая, что

$$
v_s^2 = v_y^2 + v_z^2
$$

$$
\frac{\partial }{\partial s} = \frac{\partial y}{\partial s}\frac{\partial }{\partial y} + \frac{\partial z}{\partial s}\frac{\partial }{\partial z} = \frac{v_y}{v_s}\frac{\partial }{\partial y} + \frac{v_z}{v_s}\frac{\partial }{\partial z}
$$

получим выражение для производной по потоку и поперек потока в системе координат (x,y,z):

$$
\frac{\partial}{\partial x} = \left( \frac{v_x^2}{\mathbf v^2} \frac{\partial }{\partial x} + \frac{v_x v_y}{\mathbf v^2} \frac{\partial }{\partial y} + \frac{v_x v_z}{\mathbf v^2} \frac{\partial }{\partial z} \right) + \left( \frac{v_y^2 + v_z^2}{\mathbf v^2}\frac{\partial}{\partial x} - \frac{v_x v_y}{\mathbf v^2}\frac{\partial }{\partial y} - \frac{v_x v_z}{\mathbf v^2}\frac{\partial }{\partial z}\right)
$$

Циклической перестановкой можно получить аналогичные выражения для производных по другим направлениям. 

Введем обозначения:

$$
\nabla_{||} = \left(  \frac{v_x^2}{\mathbf v^2} \frac{\partial }{\partial x} + \frac{v_x v_y}{\mathbf v^2} \frac{\partial }{\partial y} + \frac{v_x v_z}{\mathbf v^2} \frac{\partial }{\partial z}, \ldots, \ldots \right) -\text{производная вдоль поткоа}
$$

$$
\nabla_{\perp} = \left(\frac{v_y^2 + v_z^2}{\mathbf v^2} \frac{\partial }{\partial x} - \frac{v_x v_y}{\mathbf v^2} \frac{\partial }{\partial y} - \frac{v_x v_z}{\mathbf v^2} \frac{\partial }{\partial z}, \ldots, \ldots \right) -\text{производная поперек поткоа}
$$

$$
\nabla_{||} + \nabla_{\perp} = \nabla
$$

Можно показать, что производная поперек потока не участвует в конвекции, то есть:

$$
(\mathbf v, \nabla_{\perp}) = 0
$$

## Разложение выделенных слагаемых

Тот факт, что производная вдоль потока не участвует в конвекции, имеет непосредственное следствии. Можно показать, что первое из выделенных слагаемых можно разделить на две части, одна из которых в точности компенсируется другими источниковыми слагаемыми. 

Так, производная в слагаемое $- v_x \, \partial \omega_x /\partial x$ может быть раздела на производную вдоль потока и производную поперек потока. Составляющая этого слагаемого с производной поперек потока имеет вид: 

$$
- v_x \left(\frac{v_y^2 + v_z^2}{\mathbf v^2} \frac{\partial \omega_x}{\partial x} - \frac{v_x v_y}{\mathbf v^2} \frac{\partial \omega_x}{\partial y} - \frac{v_x v_z}{\mathbf v^2} \frac{\partial \omega_x}{\partial z} \right)
$$

Аналогичным образом, часть, связанная с производной поперек потока, может быть выделена в слагаемых $-v_y \, \partial \omega_x /\partial y $ и $-v_z\, \partial \omega_x /\partial z$. Сумма трех оказывается в точности равна нулю, так как они представляют собой компоненты выражения $(\mathbf v, \nabla_{\perp}) \omega_x$. Таким образом, часть выделенного слагаемого, не связанная с растяжением и сжатием жидких частиц, полностью компенсируется другими слагаемыми в правой части уравнения. 

Для слагаемого $\omega_x \, \partial u /\partial x$, вероятно, можно провести аналогичные рассуждения, но сделать это мне пока не удалось. Можно отметить, что, так как производная поперек потока не дает вклада в конвекцию, выполнено:

$$
rot\, (\mathbf v, \nabla_{\perp}) \mathbf v - (\mathbf v, \nabla_{\perp}) \omega = 0
$$

Но отсюда не следует что-то вроде $(\omega, \nabla_{\perp})\mathbf v = 0$, хотя для оператора Набла в его исходном виде верно:

$$
(\omega, \nabla)\mathbf v = rot\, (\mathbf v, \nabla) \mathbf v - (\mathbf v, \nabla) \omega
$$

Здесь нужно еще разбираться. Если все аккуратно расписать, то что-то возможно прояснится. 

## Заключение

В предыдущем разделе я сознательно упустил штрихи, так как мне не ясно, нужно ли раскладывать производную относительно относительно поля скорости $\mathbf v'$ или $\mathbf v$. Хотя результат от этого не меняется. Кроме того, правильно ли вообще раскладывать производную относительно поля скорости, когда скорость зависит от системы отсчета? Может быть, нужно раскладывать производную около траекторий частиц, а не линий тока? 

Есть еще такая проблема, что в жидкости могут быть критические точки, где её скорость обращается в ноль. И в случае с бегущей волной в системе отсчета, связанной с ней, такие точки как раз попадают в область генерации продольных вихрей. В них приведенные выше формулы вырождаются. 

Мне представляется возможным все сформулировать аккуратно для случая бегущей волны, так как её поле скорости стационарно в подвижной системе отсчета и линии тока в нем совпадают с траекториями частиц. Потом, может быть, удастся обобщить полученные результаты и на произвольный поток жидкости. 

Цель этой работы состоит в том, чтобы продемонстрировать потенциальную возможность разделения выделенных слагаемых на части, одна из которых связана с физическим механизмом, который за ними стоит, а для второй можно показать, что она в точности компенсируется другими слагаемыми. Кроме того, имеет смысл аккуратнее расписывать уравнения, и искать именно физический смысл за слагаемыми. Может быть после того, как слагаемые, связанная с производными поперек потока, будут удалены из уравнения, остальные слагаемые между собой окажется примерно равны, и уже нельзя будет утверждать, что выделенное слагаемое доминирует.
