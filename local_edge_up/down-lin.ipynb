# Линейная устойчивость среднего течения решения на сепаратрисе

Здесь собраны вместе все сведения по линейной устойчивости $V_{edge} = \overline{v_{edge}}$ -- средней составляющей поля скорости решения на сепаратрисе $v_{edge}$, полученного при Re=2200 в круглой трубе. Осреднение выполняется в подвижной системе отсчета, связанной с порывом ($c_f = 0.69U$, $U$ -- максимальная скорость в ламинарном течении). 

## Устойчивость среднего течения $V_{edge}$

Наложим на среднее течение малые возмущения и будем наблюдать за их эволюцией. Так как возмущения малы, их нелинейным взаимодействием можно пренебрегать. Таким образом возникает линейная задача на возмущения среднего течения. Она имеет собственные решения. Найти наиболее быстро растущее собственное решение и скорость его роста можно просто интегрируя по времени произвольное начальное возмущения до выхода на экспоненциальный рост. Программа была модифицирована так, что воспроизводит эволюцию возмущений среднего течения в рамках линеаризованных уравнений. На графике ниже изображено, как растет амплитуда возмущения со временем. Решение выходит на экспоненциальный рост с показателем экспоненты $\lambda = 0.0123 U/R$. Период линейного решения во времени оказывается равен $T = 54.2 R/U$, в то время, как период реешния на сепаратрисе близок к $T = 59.6 R/U$. 

<img src="down_lin_img/lin_amp.png">

Наиболее быстро растущее собственное решение $v'_{lin}$ воспроизводит форму пульсационной составляющей движения $v'_{edge} = v_{edge} - V_{edge}$. Как сравнивать периодические во времени пол скорости? Можно подобрать момент времени для $v'_{edge}$ и $v'_{lin}$, в которые они близки, и показать это. Можно осреднить поля скорости во времени и сравнивать средние (линейное решение отнесено к $exp(\lambda t)$, так как именно такое поле скорости периодично во времени). Если сравнивать средние, то теряется информация о том, что решение является бегущей волной, её длина и скорость перемещения, тем не менее, сперва сравним средние картины течения. 

Ниже на графике изображено, как меняется амплитуда $v'_{edge}$ и $v'_{lin}$ вдоль трубы (средний по сечению трубы и по времени их квадрат). Они практически совпадают (амплитуду линеаризованного решения я отнормировал соответствующим образом).

Ниже изображена амплитуда пульсаций в нескольких сечения трубы, отмеченных вертикальными черточками на графике выше. Верхний ряд соответствует $v'_{lin}$, второй ряд соответствует $v'_{edge}$. Они хорошо совпадают. Изображена амплитуда только продольной скорости, но и другие компоненты скорости также демонстрируют совпадение.

Приведенных картинок не достаточно ,чтобы утверждать, что пульсации являются бегущей волной, но это так. Волна бежит в низ по порыву. На всех приведенных картинках изображена только проданная компоненты скорости, но и поперечные компоненты демонстрируют совпадение. Можно подумать, какую иллюстрацию построить, чтобы явно показать, что решения являются бегущими волнами, и что они совпадают. 

## Устойчивость течения $(U_{edge},0,0)$

На линейную устойчивость было проверено также течение, в котором продольная компонента скорости взята из $V_{edge}$, а поперечные равны нулю. Дивергенция такого поля скорости не равна нулю, но это не мешает провести расчет его устойчивости. Ниже на графике изображено, как меняется амплитуда некоторого начального возмущения в этом случае. Решение входит на экспоненциальное затухание с показателем экспоненты $\lambda = - 0.0064 U/R$. Экспонента с таким показателем изображена прерывистой линией. Период возникающего решения во времени оказывается близок к $T = 81.8 R/U$, что также достаточно близко к периоду решения на сепаратрисе $T = 59.6 R/U$.


<img src="down_lin_img/zVW_amp.png">

Несмотря на то, что решение оказывается линейной устойчивым, форма возникающего собственного возмущения $v'_{Ulin}$ повторяет форму $v'_{lin}$, а значит и форму $v'_{edge}$. Для того, чтобы это продемонстрировать, ниже приведено распределение амплитуды пульсаций предельной скорости $v'_{Ulin}$ (вверху) в сравнении с $v'_{lin}$ (внизу) в нескольких сечения трубы. Они неплохо совпадают. Такое же совпадение демонстрируют и другие компоненты скорости. 

<img src="down_lin_img/VWz.vs.orig.lin.png">

Ниже мгновенное поле продольной скорости $v'_{Ulin}$ (вверху) и $v'_{lin}$ (внизу) в сечении $\theta = 0$. Они также похожи. 

<img src="down_lin_img/lin.vs.orig.th0.png">

Отдельно можно показать, как $v'_{edge}$, $v'_{lin}$ и $v'_{Ulin}$ генерируют $\Omega_x$, если нежно. В общем, получается так, что и в случае $v'_{Ulin}$ те же слагаемые дают наибольший вклад в образование продольных вихрей, что и в других случаях, однако пульсации продольной завихренности возникают в результате другого механизма и иначе согласованы по фазе с пульсациями продольной скорости.
