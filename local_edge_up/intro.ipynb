# Модельный порыв на верхней ветви решения

Было найдено решение на сепаратрисе при Re=2200 в круглой трубе. Оно оказывается локализованным в пространстве и периодическим по времени в подвижной системе отсчета. Исследуя это решение, удалось выделить и продемонстрировать основные элементы цикла его самоподдержания. Возникает вопрос, насколько общими являются полученные результаты? 

Метод Ньютона-Крылова позволяет уточнить найденное решение на сепаратрисе. Кроме того, оно может быть использовано в качестве начального приближения для решения с близким значением параметров, с которым метод Ньютона сойдется. Таким образом решение может быть продлено в пространстве параметров. Таким образом могут быть получены новые решения, на которых может быть проверены полученные результаты.

Продлевая решение на сепаратрисе в сторону уменьшения Re, удается достичь точки бифуркации, в которой возникает две ветви решения, и перейти с нижней ветви решения на верхнюю. Двигаясь вверх по Re на верхней ветви удается получить новое решение при том же Re=2200. Как и исходное решение, оно локализовано в пространстве и является периодическим по времени. 

На графике ниже приведено изменение периода по времени $T$ и скорости сноса $cf$ по мере продвижения решения по Re. Красной точкой обозначено решение на верхней ветви, исследованию которого посвящена эта работа. И $T$, и $cf$ на верхней ветви имеют меньшие значения, чем на нижней -- на приведенных графиках нижняя ветви оказывается выше верхней. Верхняя ветвь названа верхней потому, что на ней решение имеет большую амплитуду пульсаций, и, в целом, её параметры ближе к тем, которые наблюдаются в турбулентном течении.

Важно отметить, что пройти по верхней ветви прямо мне не удалось. При $Re \approx 1880$ решение имеет какую-то особенность, которую я не смог преодолеть. Продлить решение в сторону больших Re мне позволило то, что я сместился в сторону меньших nsym -- включил в число варьируемых параметров еще один. Таким образом, удалось обойти эту особенность решения и вернуться к nsym=2 при большем значении Re. При этом появилась возможность пройти эту особенность в обратном направлении, но этого мне тоже сделать не удалось. Я не знаю, что это было, но глядя на графики выше можно отметить, что до этой особенности решение меняется крайне плавно и монотонно, а после оно начинает "вилять". 

В этой работе будет исследовано решение с верхней ветви, имеющее следующие параметры:
    
    Re = 2200
    cf = 0.59
    T = 31.85
    nsym = 2
    Xmax = 120
    Im * Jm * Km = 1024 x 40 x 32
    
Величины обезразмерены, как в программе на фортране, по максимальной скорости и радиусу трубы. 

Для сравнения, параметры исходного решения на сепаратрисе, которое было найдено на той же сетке:
    
    Re = 2200
    cf = 0.69
    T = 59.61
