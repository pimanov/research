Что такое d2,d3 - написано в файле math

Здесь исследуются уравнения для скорости $u$, не осредненной по времени, а в общем случае. Но сами уравнения по времени осреднены. 

$$
d_2 = \overline{ - \frac{\partial w}{\partial x} \frac{\partial u}{\partial r}}^t, 
d_3 = \overline{ \frac{\partial v}{\partial x} \frac{\partial u}{r \partial \theta} }^t
$$

Прерывистая линия - продольная компонента завихренности ox. 

Интегрирование поперек трубы проводилось следующим образом

$$
a(x_0) = \left( \frac{1}{S} \int_{x=x_0} a^2 ds \right)^{1/2}
$$

Если не возводить в квадрат, то в силу симметрии, все интегралы 0. 

Вертикальные черточки - сечения, которые далее будут рассмотрены подробнее.

На нижнем графике для того, чтобы ориентироваться, приведены:   
-streaks - отклонение от осесимметричного течения  
-pulsations - амплитуда пульсаций   
-ox - продольная компоненты завихренности. Та же кривая, что и на предыдущем графике.   

Интегрирование по сечению проводилось тем же способом. 

Если считать, что завихренность переносится вниз по порыву, то интерпретация графика может быть следующая. Сперва, там, где продольная завихренность только формируется, амплитуда каждой d растет вместе с амплитудой завихренности. Затем вклад d3 медленно падает до 0, а вклад d2 после некоторого снижения делает второй, более значительный всплеск, но он сопровождается уже угасанием продольной завихренности. Возможно он и приводит к этому угасанию. Чтобы это выяснить, имеет смысл построить подобный график для слагаемых d2\*ox, d3\*ox, через которые пополняется энергия продольной завихренности. Эти кривые на следующем графике. 

Интересно отметить, что области, где существуют ox и d практически совпадают, но пульсации происходят в значительно более компактной области. Видимо среднее течение само достаточно эффективно производит продольную завихренность. Можно посчитать отдельно вклад среднего течения и пульсаций, что было сделано и результаты в следующем файле. 

$$
d_2 = \overline{ - \frac{\partial w}{\partial x} \frac{\partial u}{\partial r}}^t = - \frac{\partial \overline{w}^t }{\partial x} \frac{\partial \overline{u}^t }{\partial r} - \overline{\frac{\partial w}{\partial x}' \frac{\partial u}{\partial r}'}^t
$$

Для $d_3 $ аналогично. 

Вообще, кривые имеет нетривиальную форму, что говорит о том, что внутри порыва происходят достаточно сложные взаимодействие.

В действительно, в уравнениях для энергии осреднение проводится поверх умножения, и тут я потерял некоторые слагаемые, но думаю, качественно картина от этого не меняется. 

Все обозначения, как на предыдущем графике. Но интегрирование проводилось без возведения в квадрат

$$
a(x_0) = \frac{1}{S} \int_{x=x_0} a ds
$$


Предположение, что продольная компоненты завихренности ox теряет энергию через слагаемое ox\*d2, подтверждается. Слагаемое ox\*d3 дает только положительный вклад. Площадь под кривой суммы слагаемых близка к 0, что согласуется с тем, что только через эти два слагаемых энергия попадает в продольную завихренность. 

Получается, что продольная завихренность переносится вниз по порыву. При этом она формируется там, где пульсаций нет, только за счет среднего течения. Исследование отдельно вклада от пульсаций, отдельно от среднего течения, показали, что среднее течение дает только положительный вклад, а пульсации - только отрицательный. Это противоречит ожиданиям. Наверное может, но пока я об этом думал, это натолкнуло меня на некоторые соображения, они в конце этого файла.

### Та мысль, на которую меня натолкнуло рассуждение о том, верно ли получен знак d

Результаты получаются противоречивыми. Получается так, что завихренность формируется в той области, где пульсаций нет, а там, где они есть, она наоборот затухает. Возможно я просто потерял где-то знак d, и на самом деле завихренность рождается в передней части порыва, в присутствии пульсаций, а затем сносится назад, и затухает там. 

Это можно было бы просто проверить, установив, в какую сторону конвекция переносит завихренность, например, глядя на последний график. Если основной вихрь находится в области между черной линией и центром трубы, то завихренность переносится вперед, и было посчитано верно. Если он находится с другой стороны от черной линии, то все наоборот.

Но основная завихренность оказывается с одной стороны от линии в передней части порыва, и с другой стороны в задней. Конвекция выносит завихренность из центра порыва в обе стороны. Там же, где она достигает максимума, завихренность вообще покоится относительно порыва. Проблема в том, что с одной стороны порыва положительный вклад d, с другой - отрицательный, а через центр порыва завихренность как-будто не переносится. Объяснить происходящее можно было бы так, что там, где завихренность покоится, даже небольшой вклад d дает большое значение ox, так как она накапливается. А в передней и задней части порыва, какой бы вклад не был, ох все равно выносится из него. Если так, то через центр порыва переноса ox практически нет. 

Так можно объяснить формирование застойной зоны с высоким значением ox в центре порыва, но не формирование полос. Полосы вытягиваются там, где поток уносит завихренность. Соответственно источник полос там, где вклад d наибольший. Максимум d в задней части порыва, где завихренность выносится назад и формируются вытянутые полосы сзади. Но тогда ничто не производит полосы в передней части порыва, и не ясно, как формируются полосы там, где наблюдаются пульсации.

Все эти рассуждения очень условный, но мне кажется, что прояснить ситуацию поможет более детальное изучение вклада d. В частности, можно отдельно считать его вклад в той области, в которой жидкость движется быстрее порыва, и отдельно там, где медленнее. 

Чтобы утверждать, что знак d действительно получен верно, нужно еще поработать, но сейчас я вижу косвенное подтверждение этого в том, что в передней части порыва граница отрицательной и положительно завихренности ложится точно на черную линию, что говорит о том, что ох попала туда только за счет конвекции, и на месте не генерируется. В задней части порыва, где генерация есть, такого не наблюдается.
