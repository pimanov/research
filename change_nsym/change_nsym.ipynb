# Попытка увеличить nsym на верхней ветки методом Ньютона 

Можно предположить, что на верхней ветви решения, чтобы периодическое решение продолжало существовать с увеличением Re, нужно также увеличивать nsym, чтобы уменьшить размер расчетной области. Здесь описана попытка это сделать. 

$$
\vec v(x, r, \theta) = \vec v(x,r,\theta + 2\pi / n_{sym})
$$

На следующем графике сразу приведены результаты эксперимента. 

<img src="fig1.png">

*Рис.1. Синяя линия соответствует решениям, полученным раньше, при nsym = 2. Красная линия - решения, полученные в этом эксперименте при последовательном увеличении nsym. Слева - зависимость периода по времени T от Re. Справа - зависимость скорости снова cf от Re. Красная начинается при Re=1600 на верхней ветви решения (хотя тут она и внизу). Заканчивается она тоже при Re~1600, но это случайность.*

Я решил начать не с последнего найденного решения на верхней ветви (Re=1760), а с решения с меньшим Re, так как, наверное, последнее решение находится на границе существования решений, и каждая итерация его уточнения дается с большим трудом. Начав с решения с меньшим Re, увеличивая nsym, возможно вообще удастся не приближаться к границам существования решений, и пройти этот путь быстрее. 

Я начал с Re=1600. Жестко задавал nsym, а (Re, T, cf) определялись уже внутри метода Ньютона. T - период по времени, cf - скорость сноса. Так как определяются три параметра, а не два, как раньше, то решение видимо принадлежит уже не однопараметрическому, а двупараметрическому множеству - сущеcтвует значительный произвол в определении параметров. 

Можно предположить, что если с увеличением Re, для существования решения необходимо и увеличение nsym, то, вырабатываемый в результате работы метода Ньютона Re будет увеличиваться с увеличением nsym, и решение пройдет по этому коридору само. Судя по результату, этого не произошло. 

В данном случае приближение к каждому новому решение было получено линейной интерполяцией двух предыдущих. Это приводит к тому, что при поиске нового решения существует некоторая инерция. Первые два решения оказались такими, что при большем nsym число Рейнольдса оказалось меньше. Линейная интерполяция сохранила эту тенденцию до самого конца, пока траектория не уперлась в границу существования решений. Эту точку видно как на рис.1, так и на рис.2. В ней наблюдается резкий излом красной линии, в этот момент сходимость резко ухудшилась. 

<img src="fig2.png">

*Рис.2. Красные точки - решения, найденные при увеличении nsym. Всего 141 решение. По мере увеличения nsym, другие параметры сперва менялись монотонно, затем при nsym ~ 2.27 на всех кривых видно резкий излом, в этот момент сходимость резко ухудшается, шаг по Re резко увеличивается. Вероятно это связано с тем, что решение попало на границу существования решений. После этого было найдено еще несколько решений на границе, но в итоге алгоритм вообще перестал сходиться.*

Видимо, реализованный метод Ньютона не чувствует границы существования решений, и нельзя рассчитывать, что если эта граница близко, то он будет сам корректировать направление, чтобы в нее не попасть. В общем, продвинуться вверх по Re и приблизиться к nsym=3 не удалось, но это уже большой прогресс, так как nsym удалось изменить на 0.25, а в прошлый раз это было не более 0.1. 

Мне кажется, есть смысл попробовать еще раз, стартуя с большего Re, например 1700. Проследить, что с самого начала с увеличением nsym число Рейнольдса тоже растет, тогда линейная интерполяция будет поддерживать это направление. Возможно стоит вообще отказаться от интерполяции, тогда у алгоритма не будет информации о предыдущем решении и он будет более гибким. Можно не только nsym задавать жестко, но и Re, но проблема остается, так как неизвестно, как они должны быть связаны. 

Эти расчеты заняли около двух дней времени на стационарном 4-ядерном компьютере. Было найдено всего 141 решение, все они сохранены на диске. 

Я здесь использую представление, что есть граница существования решений. Так ли это на самом деле, я не знаю.
